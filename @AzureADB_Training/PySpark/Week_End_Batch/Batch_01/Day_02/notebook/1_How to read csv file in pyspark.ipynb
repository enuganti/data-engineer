     df = spark.read.csv("Advertising/advertising.csv", header=True, inferSchema=True)

- **No header** assumed.
- Column names become **_c0, _c1, _c2, â€¦**

```python
df_default = spark.read.csv("/Volumes/@azureadb/pyspark/timestamp/to_timestamp.csv")
display(df_default)
```

##### 4) Provide schema explicitly

- **Fastest & safest**.
- **Prevents wrong type inference**

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DateType

schema = StructType([
    StructField("S.No", IntegerType(), True),
    StructField("Status", StringType(), True),
    StructField("Start_Date", StringType(), True),
    StructField("End_Date", StringType(), True),
    StructField("Expiration_Date", StringType(), True),
    StructField("Last_Date", StringType(), True),
])

df_explicit = spark.read \
                   .option("header", "true") \
                   .schema(schema) \
                   .csv("/Volumes/@azureadb/pyspark/timestamp/to_timestamp.csv")

display(df_explicit)
```

##### 9) Handle quotes and escape characters
- Useful when **text columns contain commas**.

**Explanation:**
- **Outer quotes (")** â†’ define field boundaries
- **Double double-quotes ("")** â†’ represent a literal " inside the value

**Best Practice**
-  Use **quote** when text contains **delimiters**.
-  Combine with **escape** if **quotes appear inside text**.

```python
df_quote_escape = spark.read \
                       .option("header", "true") \
                       .option("quote", '"') \
                       .option("escape", '"') \
                       .csv("/Volumes/@azureadb/pyspark/training/read_csv/delimiter_quote_escape.txt")

display(df_quote_escape)
```
