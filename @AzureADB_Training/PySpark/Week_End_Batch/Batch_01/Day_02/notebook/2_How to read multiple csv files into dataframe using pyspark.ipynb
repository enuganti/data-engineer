##### 1) How to read all CSV files with different schemas / columns?

      df = spark.read.option("header", "true").csv("/Volumes/@azureadb/pyspark/timestamp/*.csv")

- Spark uses the **schema** from the **first file** it reads.
- **Columns** that **do not exist** in that schema are **dropped**.

```python
df_all = spark.read.option("header", "true").csv("/Volumes/@azureadb/pyspark/timestamp/*.csv")
display(df_all)
```

```python
df1 = spark.read.csv("/Volumes/@azureadb/pyspark/timestamp/timestamptodate.csv", header=True, inferSchema=True)
display(df1)
print("List of Columns: ", df1.columns)
print("\nNumber of Columns: ", len(df1.columns))
```

```python
df2 = spark.read.csv("/Volumes/@azureadb/pyspark/timestamp/to_timestamp.csv", header=True, inferSchema=True)
display(df2)
print("List of Columns: ", df2.columns)
print("\nNumber of Columns: ", len(df2.columns))
```

**Result:**
- Spark **picked** the **10-column** schema.
- The **extra 2 columns** from the **second file** are **ignored**.
