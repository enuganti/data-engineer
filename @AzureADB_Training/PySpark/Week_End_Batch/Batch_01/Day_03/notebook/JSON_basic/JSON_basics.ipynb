**JSON (JavaScript Object Notation)**
- built on **key-value** pairs and ordered **lists (arrays)**.
- Its basic structures are:
  - **objects ( {} with key:value pairs)**
  - **arrays ([] with comma-separated values)**
  - **strings**
  - **numbers**
  - **booleans**
  - **null**
  - **nested objects**

**Key Concepts:**

**Objects:**
- Data stored as **key-value pairs**, like `{"name": "Alice", "age": 30}`.

**Arrays:**
- Ordered lists of values, like `["apple", "banana"]`.

**Data Types:**
- Strings (in double quotes)
- numbers
- booleans (true/false)
- null
- objects
- arrays.

**Syntax:**
- Keys and string values are in double quotes;
- pairs separated by commas;
- no comma after the last item.
