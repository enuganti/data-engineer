##### getItem()
- used to **extract an element** from an **ArrayType or MapType** column.

| Use case                    | Recommended    |
| --------------------------- | -------------- |
| Arrays may be empty         | ✅ `get()`      |
| You want SQL-style          | `element_at()` |
| Arrays guaranteed non-empty | `getItem()`    |

- If index `doesn’t exist` → result will be `null`.

**a) getItem()**
- Extracting an element from an Array column
- `Arrays` guaranteed `non-empty`.

     # empty arrays
     ("Mumbai", [], [], [], [])

     col("Technology").getItem(0)
     # It tries to access index 0 of an array that has 0 elements.

     # error
     [INVALID_ARRAY_INDEX] The index 0 is out of bounds. The array has 0 elements.

**Method 01:**
- use `getItem()` with a **0-based integer index** to `access elements` from an **array**.

```python
# getItem() works only when all rows have same number of elements in an array
# "Technology" column has same number of elements in each row
df_arr_str_meth01 = (df_arr_str \
    .withColumn("get_Tech_01", col("Technology").getItem(0))   # Get first index
    .withColumn("get_Tech_02", col("Technology").getItem(1))   # Get second index
    .withColumn("get_Tech_03", col("Technology").getItem(2))   # Get third index
    .withColumn("get_Tech_04", col("Technology").getItem(3))   # Get fourth index
    .withColumn("get_Tech_05", col("Technology").getItem(4))   # Get fifth index
    .withColumn("get_Tech_06", col("Technology").getItem(5))   # Get six index
    .select("Technology", "get_Tech_01", "get_Tech_02", "get_Tech_03", "get_Tech_04", "get_Tech_05", "get_Tech_06"))

display(df_arr_str_meth01)
df_arr_str_meth01.printSchema()
```

**Method 02:**
- **square bracket syntax** -> `Pythonic square bracket notation ([])`

```python
# getItem() works only when all rows have same number of elements in an array
# "Technology" column has same number of elements in each row
df_arr_str_meth02 = (df_arr_str \
    .withColumn("get_Tech_01", col("Technology")[0])   # Get first index
    .withColumn("get_Tech_02", col("Technology")[1])   # Get second index
    .withColumn("get_Tech_03", col("Technology")[2])   # Get third index
    .withColumn("get_Tech_04", col("Technology")[3])   # Get fourth index
    .withColumn("get_Tech_05", col("Technology")[4])   # Get fifth index
    .withColumn("get_Tech_06", col("Technology")[5])   # Get six index
    .select("Technology", "get_Tech_01", "get_Tech_02", "get_Tech_03", "get_Tech_04", "get_Tech_05", "get_Tech_06"))

display(df_arr_str_meth02)
df_arr_str_meth02.printSchema()
```

##### 2) MapType column
- `Access value by key`.
- `Extracting a value` from a `Map column`.

```python
from pyspark.sql.types import MapType, StringType

data = [
    (1, {"name": "Suresh", "city": None, "age": 30, "salary": 100000, "designation": "Data Scientist"}),
    (2, {"name": "Ravi", "city": "Hyderabad", "age": 35, "salary": 120000, "designation": "Data Engineer"}),
    (3, {"name": None, "city": "Mumbai", "age": 28, "salary": 80000, "designation": "Data Analyst"}),
    (4, {"name": "Priya", "city": "", "age": 32, "salary": 96500, "designation": "Admin"}),
    (5, {"name": "Prakash", "city": "Noida", "age": 38, "salary": 75600, "designation": ""}),
    (6, {"name": "", "city": "Nasik", "age": 34, "salary": None}),
    (7, {})
]

df_map = spark.createDataFrame(data, ["id", "info"])
display(df_map)
df_map.printSchema()
```

**Method 01:**
- Use `getItem()` with the **string key** to retrieve the corresponding **value** from a **map column**.

      df["map_col"].getItem("key1")
      col("map_col").getItem("key1")

```python
df_map_key_meth01 = df_map.select(
    col("info"),
    col("info").getItem("name").alias("name"),
    col("info").getItem("city").alias("city"),
    col("info").getItem("age").alias("age"),
    col("info").getItem("salary").alias("salary"),
    col("info").getItem("designation").alias("designation")
)

display(df_map_key_meth01)
```

**Method 02:**
- **square bracket syntax** -> `Pythonic square bracket notation ([])`
- `df["map_col"]["key1"]`

```python
df_map_key_meth02 = df_map.select(
    col("info"),
    col("info")["name"].alias("name"),
    col("info")["city"].alias("city"),
    col("info")["age"].alias("age"),
    col("info")["salary"].alias("salary"),
    col("info")["designation"].alias("designation")
)

display(df_map_key_meth02)
```
