##### withColumnRenamed()
- Column `renamed permanently` in the DataFrame.
- Returns a **new DataFrame** by **renaming an existing column**.
- Because PySpark DataFrames are **immutable data structures**, this function **does not modify the original DataFrame** in place; instead, it outputs a fresh DataFrame containing the **updated schema**.

**Syntax**

     df.withColumnRenamed(existingColumnName, newColumnName)

**existingColumnName:**
- A **string** containing the **current name** of the **column** you want to **change**.

**newColumnName:**
- A **string** representing the desired **new name** for that column

##### 1) How to Rename a single column?

```python
from pyspark.sql.functions import col
```

```python
# Sample data
data = [(101, "Girish", "Kumar", "Bangalore", 26),
        (102, "Ramesh", "Rathod", "Chennai", 29),
        (103, "Somesh", "Sekar", "Hyderabad", 32),
        (104, "Pradeep", "Rao", "Pune", 25),
        (105, "Kamal", "Prathap", "Mumbai", 27),
        (106, "Kishore", "Kumar", "Nasik", 34),
        (107, "Hari", "Shetty", None, 35),
        (108, "Basha", "Azmath", None, 24)
       ]

columns = ["EmpId", "first_name", "last_name", "City", "Age"]

# Create DataFrame
df = spark.createDataFrame(data, columns)
display(df)
df.printSchema()
```

```python
# Rename 'first_name' to 'emp_first_name'
df_renamed = df.withColumnRenamed("first_name", "emp_first_name")
df_renamed.display()

print("Original Schema: ", df.printSchema())
print("Schema after rename: ", df_renamed.printSchema())
```

```python
df_dict_sgl = df.withColumnsRenamed({"first_name": "emp_first_name"})
display(df_dict_sgl)

print("Original Schema: ", df.printSchema())
print("Schema after rename: ", df_dict_sgl.printSchema())
```

##### 2) How to Rename multiple columns (chaining)?

```python
df_multi = df_renamed \
    .withColumnRenamed("EmpId", "emp_id") \
    .withColumnRenamed("last_name", "emp_last_name") \
    .withColumnRenamed("Age", "emp_age")

df_multi.display()

print("Original Schema: ", df_renamed.printSchema())
print("Schema after rename: ", df_multi.printSchema())
```

```python
df_dict_mltl = df.withColumnsRenamed({"EmpId": "emp_id", "last_name": "emp_last_name", "Age": "emp_age", "first_name": "emp_first_name"})

display(df_dict_mltl)

print("Original Schema: ", df.printSchema())
print("Schema after rename: ", df_dict_mltl.printSchema())
```

```python
# Rename columns using dictionary and .withColumnRenamed()

mapping = {
    "EmpId": "emp_id",
    "first_name": "emp_first_name",
    "last_name": "emp_last_name",
    "City": "emp_city",
    "age": "emp_age",
}

df_modern = df.withColumnsRenamed(mapping)
display(df_modern)

print("Original Schema: ", df.printSchema())
print("Schema after rename: ", df_modern.printSchema())
```

##### 4) Rename columns with special characters / spaces
- Very common in `CSV ingestion`.

```python
# Sample data
data = [(101, "Girish", "Bangalore", 26),
        (102, "Ramesh", "Chennai", 29),
        (103, "Somesh", "Hyderabad", 32),
        (104, "Pradeep", "Pune", 25),
        (105, "Kamal", "Mumbai", 27)
       ]

columns = ["emp id", "emp name", "City", "Age$"]

# Create DataFrame
df1 = spark.createDataFrame(data, columns)
display(df1)

df1.printSchema()
```

```python
df_clean = df1 \
    .withColumnRenamed("emp id", "Emp_Id") \
    .withColumnRenamed("emp name", "Emp_Name") \
    .withColumnRenamed("Age$", "Age")

df_clean.display()

print("Original Schema: ", df1.printSchema())
print("Schema after rename: ", df_clean.printSchema())
```

##### 6) Rename columns using col().alias() with a Python dictionary and Python list comprehension

```python
rename_dict = {
    "emp id": "Emp_Id",
    "emp name": "Emp_Name",
    "Age$": "Emp_Age"
}

# df.select("emp id").alias("Emp_Id")

df_renamed = df1\
  .select([col(c).alias(rename_dict.get(c, c)) for c in df1.columns])

display(df_renamed)

print("Original Schema: ", df1.printSchema())
print("Schema after rename: ", df_renamed.printSchema())
```

##### 7) Rename columns after join

```python
emp_data = [
    (101, "Ravi", 10),
    (102, "Anita", 20),
    (103, "Suresh", 10),
    (104, "Kiran", 30),
    (105, "Avinash", 35)
]

emp_df = spark.createDataFrame(emp_data, ["emp_id", "emp_name", "dept_id"])
emp_df.display()
```

```python
dept_data = [
    (10, "Engineering"),
    (20, "HR"),
    (30, "Finance"),
    (40, "Sales"),
    (50, "Marketing")
]

dept_df = spark.createDataFrame(dept_data, ["dept_id", "dept_name"])
dept_df.display()
```

```python
emp_df.alias("e") \
    .join(dept_df.alias("d"), "dept_id") \
    .display()
```

- In PySpark, the default join type is: `INNER JOIN`.
- It returns only **matching rows** from both DataFrames based on **dept_id**.

**Other Join Types in PySpark**

| Join Type            | Syntax        |
| -------------------- | ------------- |
| Inner Join (default) | `"inner"`     |
| Left Join            | `"left"`      |
| Right Join           | `"right"`     |
| Full Join            | `"full"`      |
| Left Semi Join       | `"left_semi"` |
| Left Anti Join       | `"left_anti"` |
| Cross Join           | `"cross"`     |

     emp_df.alias("e") \
         .join(dept_df.alias("d"), "dept_id") \
         .display()
             (or)
     emp_df.alias("e") \
         .join(dept_df.alias("d"), "dept_id", "inner") \
         .display()        

```python
emp_df.alias("e") \
    .join(dept_df.alias("d"), "dept_id") \
    .select("e.*", "d.dept_name") \
    .withColumnRenamed("dept_name", "department_name") \
    .display()
```

##### 8) How to Rename ALL columns to uppercase?

```python
df_upper = df1
for c in df1.columns:
    df_upper = df_upper.withColumnRenamed(c, c.upper())

df_upper.display()

print("Original Schema: ", df1.printSchema())
print("Schema after rename: ", df_upper.printSchema())
```

##### 9) How to Rename columns with prefix / suffix?

```python
# df_prefixed = df1
df_prefixed = df_renamed
for c in df_renamed.columns:
    df_prefixed = df_prefixed.withColumnRenamed(c, f"src_{c}")

df_prefixed.display()

print("Original Schema: ", df_renamed.printSchema())
print("Schema after rename: ", df_prefixed.printSchema())
```

```python
df_prefixed = df1
for c in df1.columns:
    df_prefixed = df_prefixed.withColumnRenamed(c, f"{c} new")

df_prefixed.display()

print("Original Schema: ", df1.printSchema())
print("Schema after rename: ", df_prefixed.printSchema())
```

```python
df_prefixed = df1
for c in df1.columns:
    df_prefixed = df_prefixed.withColumnRenamed(c, f"Src_{c}_New")

df_prefixed.display()

print("Original Schema: ", df1.printSchema())
print("Schema after rename: ", df_prefixed.printSchema())
```

```python
df_alias = df1.select(col("emp id").alias("EmpId"), col("emp name").alias("EmpName"), col("City"), col("Age$").alias("Age"))

display(df_alias)

print("Original Schema: ", df1.printSchema())
print("Schema after rename: ", df_alias.printSchema())
```

**withColumnRenamed() vs alias()**

| withColumnRenamed    | alias                    |
| -------------------- | ------------------------ |
| Permanent rename     | Temporary rename         |
| Works on DataFrame   | Used in `select()`       |
| One column at a time | Multiple columns at once |
| Schema change        | No schema change         |
