#### How to `Create dataframe` with `array column`?

##### Array

- Array function is used to create a **new column** of **array type** by **combining two columns**.
- All input columns must have the **same data type**.
- **All elements** of ArrayType should have the **same type of elements**.
- when we work with **json** data, it is very common to get **ArrayType columns**.

##### 1) `array column` (using `Python list of tuple`)

```python
# Create dataframe with array column
data = [(1, "Suresh", [".net", "Python", "Spark", "Azure"]),
        (2, "Ramya", ["java", "PySpark", "AWS"]),
        (3, "Rakesh", ["ADF", "SQL", None, "GCC"]),
        (4, "Apurba", ["C", "SAP", "Mainframes"]),
        (5, "Pranitha", ["COBOL", "DEVOPS"]),
        (6, "Sowmya", ["ABAP"]),
        (7, "Anand", None),
        (8, "Sourabh", [])]

schema = ["id", "Name", "skills"]

df = spark.createDataFrame(data=data, schema=schema)
# display(df)
df.show(truncate=False)
# df.show(truncate=True)
df.printSchema()
print("Number of Rows:", df.count())
```

- When you pass a Python list (e.g., `[.net, Python, Spark, Azure]`) as a column value to `spark.createDataFrame`, Spark `automatically infers` the column type as **ArrayType(StringType())** if you `do not specify a schema`.
- This is because Spark's DataFrame API can `infer complex types` like `arrays and structs` from native Python objects such as `lists and tuples`.

##### 2) `Array column` using `array()` function

```python
from pyspark.sql.functions import array, col
```

```python
data = [
    (1, "Ravi", "kumar", ["Maths", "Science", "English"]),
    (2, "Anita", "Anand", ["Biology", "Chemistry", "Social"]),
    (3, "senthil", "kumar", ["Physics"]),
    (4, "Rakesh", "kumar", ["Physics", "Computer"]),
    (5, "Lokesh", "Royal", ["Biology", "Chemistry"]),
    (6, "Sourabh", "Saxena", ["English", "History"]),
    (7, "Sathya", "kumar", ["Literature", "Sports", "English"]),
    (8, "Ganesh", "Sabhar", ["Geography", "Language"]),
    (9, "Harish", "kumar", ["Arts", "History"])
]

df_arry = spark.createDataFrame(data, ["id", "first_name", "last_name", "subjects"])
display(df_arry)
df_arry.printSchema()
```

```python
# create array from multiple columns
df_mlt = df_arry.withColumn(
    "subject_array",
    array(col("first_name"), col("last_name"))
)

display(df_mlt)
```

##### 4) Using explicit schema

```python
from pyspark.sql.types import *

schema = StructType([
    StructField("employee_id", StringType()),
    StructField("employee_name", StringType()),
    StructField("experience", IntegerType()),
    StructField("designation", StringType()),
    StructField("skills", ArrayType(StringType()))
])

data = [
    ("E101", "Kamal", 2, "Engineer", ["Python", "Spark", "SQL"]),
    ("E102", "Suresh", 3, "Architect", ["Python", "Spark", "SQL"]),
    ("E103", "Ravi", 4, "Developer", ["Scala", "Spark", "SQL"]),
    ("E104", "Priya", 5, "Testing", ["PySpark", "SparkSQL", "SQLServer"]),
    ("E105", "Rajesh", 6, "Data Analyst", ["VS Code", "Airflow", "Databricks"]),
    ("E106", "Suman", 7, "SrManager", ["Git", "Confluence", "SQL"]),
    ("E107", "Swapna", 8, "Manager", ["VB", "AWS", "Azure"]),
    ("E108", "Sud", 9, "Lead", ["GCC", "Glue", "Databricks", "SQL"]),
    ("E109", "Ravi", 10, "Tester", ["Oracle", "Testing", "PowerBi"]),
    ("E110", "Raj", 11, "SrTester", ["Tableau", "ADLS", "Databricks"])
]

df_expl = spark.createDataFrame(data, schema)
display(df_expl)
```
