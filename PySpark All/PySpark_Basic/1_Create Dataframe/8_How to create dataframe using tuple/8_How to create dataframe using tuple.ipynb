#### How to create dataframe using tuple?

**Golden Rule**

| Row Data            | Schema                     |
| ------------------- | -------------------------- |
| Tuple with N values | Schema must have N columns |
| Tuple with 4 values | Schema must have 4 columns |

##### Define column names

     # This is just a Python list.
     columns = ["first_name", "age"]

- It defines the **column names** you want in your DataFrame.
  - `first_name` → `name of the person`
  - `age` → `age of the person`

##### Create the data

     # This is a list of tuples
     data = [("bob", 47), ("li", 23), ("leonard", 51)]

- `Each tuple` = `one row`
- `Each value` inside the `tuple` matches the `column order`

| first_name | age |
| ---------- | --- |
| bob        | 47  |
| li         | 23  |
| leonard    | 51  |

**1) Flat tuple**
- `Separate columns`

```python
data = [("Arnab", 31, 10, 30000),
        ("Henry", None, 8, 25000),
        ("Jayesh", 29, 4, None),
        ("Jagadish", 24, 3, 20000),
        ("Kamalesh", 21, None, 15000),
        ("Lepakshi", 23, 2, 18000),
        ("Anand", None, None, 40000),
        ("NULL", 34, 10, 38000),
        ("NULL", 36, None, None)]

columns = ["EmpName", "Age", "Experience", "Salary"]

df_tuple_01 = spark.createDataFrame(data, columns)
display(df_tuple_01)
```

**Spark rule for createDataFrame**

| Data Row Type   |     How Spark Interprets         |
| --------------- | -------------------------------- |
| `Tuple / List`  | `Each element` → `one column`    |
| `Dictionary`    | `Keys → columns`                 |

| Data Structure     | Result in DataFrame |
| ------------------ | ------------------- |
| Tuple inside tuple | StructType column   |
| Flat tuple         | Separate columns    |

**2) Tuple inside tuple**
- `StructType column`

```python
data = [(("Arnab", 31, 10, 30000),),
        (("Henry", None, 8, 25000),),
        (("Jayesh", 29, 4, None),),
        (("Jagadish", 24, 3, 20000),),
        (("Kamalesh", 21, None, 15000),),
        (("Lepakshi", 23, 2, 18000),),
        (("Anand", None, None, 40000),),
        (("NULL", 34, 10, 38000),),
        (("NULL", 36, None, None),)]

columns = ["EmpName", "Age", "Experience", "Salary"]

df_tuple_02 = spark.createDataFrame(data, columns)
display(df_tuple_02)
```

##### 3) Why schema (column names) is required here?

**Spark does NOT know:**
- What the `3 values` inside the **tuple** represent?
- Whether they are **first/middle/last name?**
- Whether the `tuple` should be **flattened or kept** as a `single column`.

- Tuples **do NOT** carry **column names**

      # Unlike `dictionaries`:
      {'first':'Rahul', 'middle':'', 'last':'Prakash'}

      # A tuple:
      ("Rahul","","Prakash")

      ❌ No keys

      ❌ No labels

      ❌ No structure metadata
      # So Spark cannot guess column names.

**Spark `only infers data types, not structure`**
- Spark `can infer`:
  - `1 → integer`
  - `"Rahul" → string`

- But Spark `cannot infer`:

  - Should this tuple be:
    - `A single column?`
    - `Three separate columns?`
    - `A struct?`

- So Spark requires you to decide by `passing schema`.

```python
from pyspark.sql.types import IntegerType, StringType, ArrayType, StructField, StructType

schema = StructType([
    StructField('id', IntegerType(), True),
    StructField('Name', StructType([
        StructField('firstname', StringType(), True),
        StructField('middlename', StringType(), True),
        StructField('lastname', StringType(), True)
    ]))
    ])

df_schema = spark.createDataFrame(data, schema)
display(df_schema)
df_schema.printSchema()
```
