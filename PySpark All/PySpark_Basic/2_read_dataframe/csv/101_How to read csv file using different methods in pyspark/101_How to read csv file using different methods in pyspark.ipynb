#### Different methods to read .csv files?

#### content
- Basic way (default options)
- Read CSV with header
- Infer schema automatically
- Provide schema explicitly
- Using spark.read.format("csv")
- Using options with key-value pair

     # Method 01
     df = spark.read.csv("Advertising/advertising.csv", header=True, inferSchema=True)

     # Method 02
     df = spark.read.option("header", True).csv("Advertising/advertising.csv")
 
     # Method 03
     df = spark.read.format('csv').option(key='header', value=True).load(path='dbfs:/FileStore/data/Employee.csv')

##### 1) Basic way (default options)

- While reading **csv** file by **default** every column detects as **string data type**.
- **No header** assumed.
- Column names become **_c0, _c1, _c2, …**

```python
df_default = spark.read.csv("/Volumes/azureadb/pyspark/timestamp/to_timestamp.csv")
display(df_default)
df_default.printSchema()
```

##### 2) Read CSV with header

- Uses the **first row** as **column names**.

      # Method 01
      df = spark.read.csv("/Volumes/@azureadb/pyspark/timestamp/to_timestamp.csv", header=True)

      # Method 02
      df = spark.read.option("header", "true").csv("/Volumes/@azureadb/pyspark/timestamp/to_timestamp.csv")

```python
df_header = spark.read.option("header", "true").csv("/Volumes/azureadb/pyspark/timestamp/to_timestamp.csv")
display(df_header)
df_header.printSchema()
```

##### 3) Infer schema automatically

- Spark **detects** data types like **int, double, date**.
- **Slightly slower** for **large files** (extra pass over data).

      # Method 01
      df = spark.read.csv("/Volumes/@azureadb/pyspark/timestamp/to_timestamp.csv", header=True, inferSchema=True)

      # Method 02
      df = spark.read.option("header", "true").option("inferSchema", "true").csv("/Volumes/@azureadb/pyspark/timestamp/to_timestamp.csv")

```python
df_infer = spark.read \
                .option("header", "true") \
                .option("inferSchema", "true") \
                .csv("/Volumes/azureadb/pyspark/timestamp/to_timestamp.csv")
                
display(df_infer)
df_infer.printSchema()
```

```python
df_infer.columns
```

##### 4) Provide schema explicitly

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

# Use StringType for date columns if the CSV contains dates in default string format (e.g., '2025-12-26')
schema = StructType([
    StructField("S.No", IntegerType(), True),
    StructField("Status", StringType(), True),
    StructField("Start_Date", StringType(), True),
    StructField("End_Date", StringType(), True),
    StructField("Expiration_Date", StringType(), True),
    StructField("Last_Date", StringType(), True),
    StructField("Input_Start_Date", StringType(), True),
    StructField("Input_End_Date", StringType(), True),
    StructField("Delivery_Start_Date", StringType(), True),
    StructField("Delivery_End_Date", StringType(), True),
    StructField("Delivery_Last_Date", StringType(), True),
    StructField("Pricing_Date", StringType(), True),
])
```

```python
df_explicit = spark.read \
                   .option("header", "true") \
                   .schema(schema) \
                   .csv("/Volumes/@azureadb/pyspark/timestamp/to_timestamp.csv")

display(df_explicit)
```

##### 5) Using spark.read.format("csv")

- Useful when **switching** between **formats dynamically**.

```python
# Chaining multiple options
df_format = spark.read \
                 .format("csv") \
                 .option("header", "true") \
                 .option("inferSchema", "true") \
                 .load("/Volumes/azureadb/pyspark/timestamp/to_timestamp.csv")

display(df_format)
df_format.printSchema()
```
