#### How to read csv file in pyspark using delimiter?

##### Using inferschema and delimiter

**Common separators:**
- `, → comma (default)`
- `| → pipe`
- `\t → tab`
- `; → semicolon`

**a) Comma-separated CSV (default delimiter)**

```python
# Using inferschema and delimiter
df_delimiter_wo_comma = spark.read.option("inferSchema", True) \
                                  .option("header", "true") \
                                  .csv("/Volumes/azureadb/pyspark/training/csv/read_csv/delimiter_comma.txt")

display(df_delimiter_wo_comma)
df_delimiter_wo_comma.printSchema()
```

```python
# Using inferschema and delimiter
df_delimiter_comma = spark.read.option("inferSchema", True) \
                         .option("header", "true") \
                         .option("delimiter", ",") \
                         .csv("/Volumes/azureadb/pyspark/training/csv/read_csv/delimiter_comma.txt")

display(df_delimiter_comma)
df_delimiter_comma.printSchema()
```

**b) Pipe-separated CSV file**

     # Method 01

```python
# Using inferschema and delimiter
df_delimiter_wo_pipe = spark.read.option("inferSchema", True) \
                         .option("header", "true") \
                         .csv("/Volumes/azureadb/pyspark/training/csv/read_csv/delimiter_pipe.txt")

display(df_delimiter_wo_pipe)
df_delimiter_wo_pipe.printSchema()
```

```python
# Using inferschema and delimiter
df_delimiter_pipe = spark.read.option("inferSchema", True) \
                              .option("header", "true") \
                              .option("delimiter", "|") \
                              .csv("/Volumes/azureadb/pyspark/training/csv/read_csv/delimiter_pipe.txt")

display(df_delimiter_pipe)
df_delimiter_pipe.printSchema()
```

**c) Tab-separated (TSV) file**
- (Columns are separated by TAB, not spaces)

| Point                   | Explanation             |
| ----------------------- | ----------------------- |
| `\t` = tab character    | Not visible in file     |
| Must specify explicitly | Spark won’t auto-detect |
| `.option("sep", "\t")` `.option("delimiter", "\t")` | Same as `delimiter`     |
| Works only at read time | No post-read effect     |

```python
# Using inferschema and delimiter
df_delimiter_wo_tab = spark.read.option("inferSchema", True) \
                                .option("header", "true") \
                                .csv("/Volumes/azureadb/pyspark/training/csv/read_csv/delimiter_tab.txt")

display(df_delimiter_wo_tab)
df_delimiter_wo_tab.printSchema()
```

```python
# Using inferschema and delimiter
df_delimiter_tab = spark.read.option("inferSchema", True) \
                             .option("header", "true") \
                             .option("delimiter", "\t") \
                             .csv("/Volumes/azureadb/pyspark/training/csv/read_csv/delimiter_tab.txt")

display(df_delimiter_tab)
df_delimiter_tab.printSchema()
```

```python
# Using inferschema and delimiter
df_delimiter_tab_sep = spark.read \
    .option("inferSchema", True) \
    .option("header", "true") \
    .option("sep", "\t") \
    .csv("/Volumes/azureadb/pyspark/training/csv/read_csv/delimiter_tab.txt")

display(df_delimiter_tab_sep)
df_delimiter_tab_sep.printSchema()
```

**d) Semicolon-separated**
- not every “CSV” is **comma-separated**.
- If your dataframe loads as **one column**, the file may be **semicolon-delimited (;)**. 

     # Fix it by setting the delimiter while reading.
     df = (spark.read.option("header", "true")
                     .option("delimiter", ";")
                     .csv(path))

```python
# Using inferschema and delimiter
df_delimiter_wo_semi = spark.read.option("inferSchema", True) \
                                 .option("header", "true") \
                                 .csv("/Volumes/azureadb/pyspark/training/csv/read_csv/delimiter_semicolon.txt")

display(df_delimiter_wo_semi)
df_delimiter_wo_semi.printSchema()
```

```python
# Using inferschema and delimiter
df_delimiter_semi = spark.read.option("inferSchema", True) \
                         .option("header", "true") \
                         .option("delimiter", ";") \
                         .csv("/Volumes/azureadb/pyspark/training/csv/read_csv/delimiter_semicolon.txt")

display(df_delimiter_semi)
df_delimiter_semi.printSchema()
```
