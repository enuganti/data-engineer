##### How to load files `before & after modified time` while reading data?

##### Modification Time Path Filters

- **modifiedBefore and modifiedAfter** are options that can be applied **together or separately** in order to achieve greater granularity over which files may load during a Spark batch query.
- Note that **Structured Streaming** file sources **don’t support** these options.

**modifiedBefore:**
- an optional timestamp to only `include files` with `modification times` occurring `before` the specified time.
- `Timestamp format` must be `ISO-8601`.

       yyyy-MM-dd'T'HH:mm:ss (e.g. 2020-06-01T13:00:00)

| File name  | Last modified |
| ---------- | ------------- |
| data1.json | 2024-01-10    |
| data2.json | 2025-03-15    |
| data3.json | 2025-05-21    |

      df = spark.read.load(
          "/Volumes/azureadb/pyspark/training/json/read_json/",
          format="json",
          modifiedBefore="2025-05-21T08:30:00"
      )

**Result:**

| File Name | Status |
|-----------|--------|
| data1.json | ✔ READ |
| data2.json | ✔ READ |
| data3.json | ❌ SKIPPED |

**modifiedAfter:**
- an optional timestamp to only include files with modification times occurring **after the specified time**.
- When a timezone option is **not provided**, the timestamps will be interpreted according to the Spark session timezone (`spark.sql.session.timeZone`).
- `Timestamp format` must be `ISO-8601`.

       yyyy-MM-dd'T'HH:mm:ss (e.g. 2020-06-01T13:00:00)

| File name  | Last modified |
| ---------- | ------------- |
| data1.json | 2024-01-10    |
| data2.json | 2025-03-15    |
| data3.json | 2025-05-21    |

      df = spark.read.load(
          "/Volumes/azureadb/pyspark/training/json/read_json/",
          format="json",
          modifiedAfter="2025-05-21T08:30:00"
      )

**Result:**

| File Name | Status |
|-----------|--------|
| data1.json | ❌ SKIPPED |
| data2.json | ❌ SKIPPED |
| data3.json | ✔ READ |

##### 1) modifiedBefore

- `Reads all JSON files`.
- Only `includes files last modified before 2050-07-01 08:30:00`.
- `Ignores files modified after` that time.

```python
df_all = spark.read \
    .load("/Volumes/azureadb/pyspark/training/json/read_json_singleline_nested_same_schema/", format="json")

display(df_all.limit(5))
print("Total Records: ", df_all.count())
```

```python
# Only load files modified before 2025-12-27T08:30:00
df_before = spark.read \
    .load("/Volumes/azureadb/pyspark/training/json/read_json_singleline_nested_same_schema/", format="json", modifiedBefore="2025-12-27T08:30:00")

display(df_before)
```

##### 3) Using modifiedAfter + modifiedBefore (date range)

**Use Case:**

      df = spark.read.load(
          "/Volumes/azureadb/pyspark/training/json/read_json/",
          format="json",
          modifiedAfter="2024-01-01T00:00:00",
          modifiedBefore="2024-12-31T23:59:59"
      )

- Read only files `modified in 2024`.
- Very useful for `incremental loads`.

```python
from pyspark.sql.types import StructField, StructType, IntegerType, StringType, LongType

# Define custom schema for the nested structure
AddSchema_Address = StructType([StructField('country', StringType(), False),
                                StructField('user', StringType(), False),
                                StructField('Location', StringType(), False),
                                StructField('Zipcode', StringType(), False)]
                              )

AddSchema_domain = StructType([StructField('Level', StringType(), False),
                               StructField('technology', StringType(), False),
                               StructField('experience', StringType(), False),
                               StructField('Age', StringType(), False)]
                              )

# Define the main schema including the nested structure
Schema = StructType([StructField('S.No', IntegerType(), False),
                     StructField('First_Name', StringType(), False),
                     StructField('Last_Name', StringType(), False),
                     StructField('source', StringType(), False),
                     StructField('description', StringType(), False),
                     StructField('input_timestamp', LongType(), False),
                     StructField('last_update_timestamp', LongType(), False),
                     StructField('country', StringType(), False),
                     StructField('Address', AddSchema_Address),
                     StructField('domain', AddSchema_domain)]
                     )
```

```python
df_after_before = spark.read \
    .load("/Volumes/azureadb/pyspark/training/json/read_json_singleline_nested_same_schema/",
            format="json",
            schema=Schema,
            modifiedAfter="2025-01-01T00:00:00",
            modifiedBefore="2025-12-31T23:59:59"
            )

display(df_after_before)
```

```python
df_after_before = spark.read \
    .load("/Volumes/azureadb/pyspark/training/json/read_json_singleline_nested_same_schema/",
            format="json",
            schema=Schema,
            modifiedAfter="2025-12-28T00:00:00",
            modifiedBefore="2026-06-20T23:59:59"
            )

display(df_after_before)
```
