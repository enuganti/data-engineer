##### How to handle `date / timestamp format` while reading csv file?

- When **reading CSV files** in PySpark, **date columns** are usually read as **string** unless you **explicitly define or convert them**.

- You can handle date formats using:

  - `inferSchema = True`
  - `schema definition / explicit schema`
  - `to_date()`
  - `to_timestamp()`  
  - `dateFormat`
  - `timestampFormat`

##### Common Date Formats

| Format       | Example    |
| ------------ | ---------- |
| `yyyy-MM-dd` | 2025-01-15 |
| `dd/MM/yyyy` | 15/02/2025 |
| `MM-dd-yyyy` | 03-20-2025 |
| `yyyy/MM/dd` | 2025/01/15 |

| Property Name | Default | Meaning | Scope |
|---------------|---------|---------|-------|
| dateFormat    | yyyy-MM-dd | Sets the **string** that indicates a **date format**. **Custom date formats** follow the formats at Datetime Patterns. This applies to date type. | read/write|
| timestampFormat | yyyy-MM-dd'T'HH:mm:ss[.SSS][XXX] | Sets the **string** that indicates a **timestamp format**. **Custom date formats** follow the formats at **Datetime Patterns**. This applies to **timestamp type**. | read/write |

##### Reading Data with a Custom Schema

     from pyspark.sql.types import StructType, StructField, IntegerType, DateType

     # Explicitly defining the schema with DateType
     schema = StructType([
         StructField("id", IntegerType(), True),
         StructField("birth_date", DateType(), True)
     ])

     # Reading the file 
     df = spark.read \
         .option("header", "true") \
         .schema(schema) \
         .option("dateFormat", "yyyy-MM-dd") \
         .csv("path/to/file.csv")

     display(df)

##### Writing Data

     df.write \
         .option("header", "true") \
         .option("dateFormat", "yyyy-MM-dd") \
         .csv("path/to/output_directory")

##### Critical Rules for Pattern Strings

If you need to change `"yyyy-MM-dd"` to match other `custom formats (e.g., "dd/MM/yyyy")`, keep these rules in mind:

- **Case Sensitivity Matters:**
  - `yyyy = 4-digit year (lowercase)`.
  - `Do not use uppercase YYYY`, which refers to the `week-based calendar year`.
  - `MM = 2-digit month (uppercase)`.
  - `Lowercase mm stands for minutes`.
  - `dd = 2-digit day of the month (lowercase)`.
 
- **Timestamps Require a Different Option:**
  - The `dateFormat` option only `handles dates (DateType)`.
  - If your column contains `hours, minutes, or seconds`, you must use `.option("timestampFormat", "yyyy-MM-dd HH:mm:ss")` instead.

- When Spark `reads a CSV`, it treats `all columns` as `StringType` by default unless:
  - You enable schema inference (inferSchema=true), or
  - You provide an explicit schema.

- dateFormat `does not automatically` convert a `CSV column to DateType`.
- `dateFormat` specifies the `pattern` used to `parse dates`.
- It `does not` change a `StringType column to DateType`.
- Use `inferSchema=true` or an `explicit DateType schema` to get `date columns instead of strings`.

```python
df_inf = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("dateFormat", "yyyy-MM-dd") \
    .csv("/Volumes/azureadb/pyspark/training/csv/dateFormat/yyyy-MM-dd_default_format.csv")

display(df_inf)
df_inf.printSchema()
```

**.option("dateFormat", "yyyy-MM-dd"):**
- is used during **data ingestion (like reading CSV or JSON files)** to tell Spark how to **interpret or write date strings**.

**Key Usage ScenariosReading Data:**
- When using **spark.read**, this option specifies the format of the **date strings** in your **source file** so Spark can **correctly convert them into a DateType column**.

**Writing Data:**
- When using **df.write**, it ensures the resulting file **stores dates** in that **specific string format**.

```python
from pyspark.sql.types import StructType, StructField, IntegerType, DateType, StringType

schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("name", StringType(), True),
    StructField("joining_date", DateType(), True)
])

df_mixed_format_schema = spark.read \
    .option("header", "true") \
    .option("dateFormat", "yyyy-MM-dd") \
    .schema(schema) \
    .csv("/Volumes/azureadb/pyspark/training/csv/dateFormat/mixed_date_format.csv")

display(df_mixed_format_schema)
df_mixed_format_schema.printSchema()
```

- `DOB` is in the format `yyyy:MM:dd`, and `Submit Time` is in the format `yyyy-MM-dd HH:mm:ss`.

```python
df_def_format = spark.read \
  .option("header", "true") \
  .option("inferSchema", "true") \
  .option("dateFormat", "yyyy:MM:dd") \
  .option("timestampFormat", "yyyy-MM-dd HH:mm:ss") \
  .csv("/Volumes/azureadb/pyspark/training/csv/dateFormat/studentDate.csv")

display(df_def_format)
df_def_format.printSchema()
```
