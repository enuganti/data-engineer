##### **`b) Using dbutils (Python)`**
- This method allows you to specify **exactly how many bytes** to **read** and return the content as a **string**.
- `dbutils.fs.head(path, maxBytes)`  
  - `print(dbutils.fs.head("/Volumes/azureadb/pyspark/training/json/read_json/singleline.json", 1050))`

**Syntax**

     dbutils.fs.head(path, maxBytes)

- The **second argument maxBytes** means:
  - **Maximum number of bytes** to read from the **start of the file**.
    - Read only the **first 5000 bytes** of the file.
    - Display that portion.
    - **NOT the entire file** (if it’s large).

**What happens if bytes are less than JSON length?**
- You’ll see **truncated JSON**.
- It may **look incomplete or broken**.

```python
sLine = dbutils.fs.head("/Volumes/azureadb/pyspark/training/json/read_json/SingleLine/singleline.json", 1050)
print(sLine)

print("\ndata type: ", type(sLine))
```

```python
print(dbutils.fs.head("/Volumes/azureadb/pyspark/training/json/read_json/SingleLine/singleline.json", 1000))
```

```python
print(dbutils.fs.head("/Volumes/azureadb/pyspark/training/json/read_json/SingleLine/singleline.json", 1013))
```

```python
print(dbutils.fs.head("/Volumes/azureadb/pyspark/training/json/read_json/SingleLine/singleline.json", 980))
```

**Why not read the whole file?**

Because:
- JSON files can be **very large**.
- Reading the **entire file** into the **notebook** output can:
  - **Slow down the notebook**.
  - Flood the output
  - Cause UI issues or timeouts
  - So Spark / Databricks forces you to specify a limit.

**How much should you use?**

| File size           | Suggested bytes           |
| ------------------- | ------------------------- |
| Small JSON          | 1,000 – 5,000             |
| Medium JSON         | 5,000 – 20,000            |
| Debugging structure | Enough to see full object |

- For **single-line JSON**, you may need a **higher value** to see the **complete line**.

##### 2) Parse Raw Head Content to Python Object
- If you need to programmatically inspect the JSON structure from the **dbutils.fs.head** output, combine it with the Python json library

        import json

        # Read the file content as a string
        raw_json = dbutils.fs.head("/path/to/your/file.json")

        # Parse the string into a Python dictionary or list
        parsed_data = json.loads(raw_json)
        print(parsed_data)

**a) SingleLine JSON**

```python
raw_json = dbutils.fs.head("/Volumes/azureadb/pyspark/training/json/read_json/SingleLine/singleline.json")
print("Raw JSON as String: \n", raw_json)
print("\ndata type: ", type(raw_json))

print("\nSplit Lines: \n", raw_json.splitlines())
print("\ndata type: ", type(raw_json.splitlines()))
```

```python
import json

# Read file content
raw_json = dbutils.fs.head("/Volumes/azureadb/pyspark/training/json/read_json/SingleLine/singleline.json")
print("data type: ", type(raw_json))

# Convert each line into a JSON object
parsed_data = [json.loads(line) for line in raw_json.splitlines()]
print("\nparsed_data: \n", parsed_data)
print("\ndata type: ", type(parsed_data))
```

##### 4) Display as a Structured Table (DataFrame)
- To view JSON data in a **tabular format**, read it into a Spark DataFrame using PySpark.

**`a) Single-line JSON (JSON Lines)`**
- Each line in the file must be a valid, self-contained JSON object.

      df = spark.read.json("/path/to/your/file.json")
      display(df)

**`b) Multi-line JSON`**
- If the file contains a single JSON object or array that spans multiple lines, you must enable the multiLine option.

       df = spark.read.option("multiLine", "true").json("/path/to/your/file.json")
       display(df)

```python
df_single = spark.read.json("/Volumes/azureadb/pyspark/training/json/read_json/SingleLine/singleline.json")
display(df_single)
```

```python
df_mlt = spark.read.option("multiLine", "true").json("/Volumes/azureadb/pyspark/training/json/read_json/MultiLine/multiline.json")

display(df_mlt)
```

```python
df_mlt_nest = spark.read.option("multiLine", "true").json("/Volumes/azureadb/pyspark/training/json/read_json/MultiLine_Nested/multiline_nested.json")

display(df_mlt_nest)
```

```python
spark.read.json("/Volumes/azureadb/pyspark/training/json/read_json/SingleLine/singleline.json") \
     .createOrReplaceTempView("json_tbl")

spark.sql("SELECT * FROM json_tbl").display()
```
