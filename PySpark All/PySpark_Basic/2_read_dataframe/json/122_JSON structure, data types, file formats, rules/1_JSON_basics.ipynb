**JSON (JavaScript Object Notation)**
- built on **key-value** pairs and ordered **lists (arrays)**.
- Its basic structures are:
  - **objects ( {} with key:value pairs)**
  - **arrays ([] with comma-separated values)**
  - **strings**
  - **numbers**
  - **booleans**
  - **null**
  - **nested objects**
  - **arrays**

```python
'{"name":"John", "age":30, "car":null}'

# The example above defines an object with 3 properties:

    "name"
    "age"
    "car"

# Each property has a value:

    "John"
    30
    null
```

##### 1) JSON Strings
- **Strings** in JSON must be written in **double quotes**.

```python
{"name":"John"} -> clean json object (since value is double quoted)
{"name":'John'} -> corrupt record (since value is single quoted)
```

**JSON Strings can contain the following backslash-escaped characters:**
- `\" – Double quote`
- `\\ – Backslash`
- `\/ – Forward slash`
- `\b – Backspace`
- `\f – Form feed`
- `\n – Newline`
- `\r – Carriage return`
- `\t – Tab`
- `\u – Trailed by four hex digits`

##### 2) JSON Numbers
- **Numbers** in JSON must be an **integer or a floating point**.

```python
{"age":30}
{ "cores": 8, "price": 6.99 }
```

```python
{
  "aString": "Some text",
  "aNumber": 10
}
```

```python
{
  "aFloatingPointNumber": 3.1415,
  "anExponentional": 8.15e6
}
```

- Represented in **base 10** with no superfluous leading zeros (e.g. 67, 1, 100).
- Include digits **between 0 and 9**.
- Can be a **negative number (e.g. -10)**
- Can be a **fraction (e.g. .5)**.
- Can also have an **exponent of 10**, prefixed by **e or E** with a **plus or minus** sign to indicate **positive or negative exponentiation**.
- **Octal and hexadecimal formats** are not supported.
- **Can not** have a value of **NaN (Not A Number) or Infinity**.

```python
{
	"number_1" : 210,
	"number_2" : -210,
	"number_3" : 21.05,
	"number_4" : 1.0E+2
}
```

##### 3) JSON Objects
- An **unordered set** of **name/value pairs** inserted between **{}** (curly braces).
- An object can contain **zero or more name/value pairs**.
- **Multiple name/value** pairs are separated by a **, (comma)**.
- Data stored as **key-value pairs**, like `{"name": "Alice", "age": 30}`.
  - **Values** in JSON can be **objects**.
      `{
      "employee":{"name":"John", "age":30, "city":"New York"}
      }`
  - **Objects as values** in JSON must follow the **JSON syntax**.

```python
# Zero key / value pair
{}

# Single key / value pair
{"age":30}

# Multiple key / value pair
{ "cores": 8, "price": 6.99 } -> Single Line JSON

# Multi Line JSON
{
	"visibility" : true,
	"popularity" : "immense",
	"id" : 210
}
```

**Nested Objects**

     {
       "id": 456,
       "name": "Joe",
       "additionalInformation": {
         "playsGuitar": true,
         "playsPiano": false
       }
     }

##### 4) JSON Arrays
- Values in JSON can be **arrays**.
- An **ordered collection of values**.
- `Begins with [ (left bracket) and ends with ] (right bracket).`
- `it’s values are separated by , (comma).`

```python
# Array as values
{"employees":["John", "Anna", "Peter"]}

{ "ports": [80, 443] }

# json array
[
  { "id": 1, "name": "Alice" },
  { "id": 2, "name": "Bob" }
]
```

```python
{
	"ids" : ["1","2","3"]
}

//or

{
"ids" : [
		{"id" : 1},
		{"id" : 2},
		{"id" : 3}
	]
}
```

##### 5) JSON Booleans
- Values in JSON can be **true / false**.

```python
{"sale":true}
{ "enabled": true, "debug": false }
```

##### 6) JSON null
- Values in JSON can be **null**.
- When there is **no value** to assign to a **key**, it can be treated as **null**.
- null value **should not be surrounded by quotes**.

```python
{"middlename":null}
```

```python
{
	"visibility" : true,
	"popularity" : null, //empty
	"id" : 210
}
```
