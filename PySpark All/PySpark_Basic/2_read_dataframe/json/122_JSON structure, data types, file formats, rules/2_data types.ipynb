**Data Types:**
- JSON supports **six core data types** categorized into **primitives** (simple values) and **complex structures**.

**Primitive Data Types**
|  data type  |  Description        |
|-------------|---------------------|
|   String    | Textual data enclosed in double quotes (e.g., "name": "Bengaluru"). |
|   Number    | Double-precision floating-point numbers (integers or decimals) without quotes (e.g., "temperature": 28.5). |
|   Boolean   | Logical values true or false, unquoted (e.g., "isRaining": false). |
|   Null      | Represents an empty or intentionally absent value, written as null. |

**Complex Data Types**
|  data type  |  Description        |
|-------------|---------------------|
|  Object     | A collection of unordered key-value pairs wrapped in curly braces {}. Keys must be strings and separated by commas. (e.g., {"user": {"id": 1, "active": true}}) |
|  Array      | An ordered, comma-separated list of values wrapped in square brackets []. Array items can be any valid JSON data type (e.g., ["apple", "banana", "cherry"])  |

JSON **values cannot** be one of the following data types:

- a function
- a date
- undefined

**JSON supports the following data types:**

| Data Type   | Example                                        |
| ----------- | ---------------------------------------------- |
| **String**  | `"name": "John"`                               |
| **Number**  | `"age": 25`, `"salary": 5000.75`               |
| **Boolean** | `"isActive": true`                             |
| **Null**    | `"middleName": null`                           |
| **Object**  | `"address": {"city":"Bangalore","pin":560001}` |
| **Array**   | `"skills": ["Python","Spark","SQL"]`           |

```python
# Example JSON containing all data types
{
  "name": "John",
  "age": 30,
  "salary": 5000.50,
  "isEmployee": true,
  "manager": null,
  "address": {
    "city": "Bangalore",
    "state": "Karnataka"
  },
  "skills": ["Python", "Spark", "SQL"]
}
```

**Mapping to PySpark Data Types**

| JSON Type        | PySpark Type             |
| ---------------- | ------------------------ |
| String           | StringType               |
| Number (integer) | IntegerType / LongType   |
| Number (decimal) | DoubleType / DecimalType |
| Boolean          | BooleanType              |
| Null             | Nullable field           |
| Object           | StructType               |
| Array            | ArrayType                |

```python
from pyspark.sql.types import *

schema = StructType([
    StructField("name", StringType()),
    StructField("age", IntegerType()),
    StructField("salary", DoubleType()),
    StructField("isEmployee", BooleanType()),
    StructField("manager", StringType()),
    StructField("address", StructType([
        StructField("city", StringType()),
        StructField("state", StringType())
    ])),
    StructField("skills", ArrayType(StringType()))
])
```
