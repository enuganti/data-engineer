#### How to read multiLine JSON files in pyspark?

**JSON Lines (JSONL)**
- By default, PySpark expects a **JSON Lines (JSONL)** format where **every single line** is an **independent**, complete JSON object.
- **JSON Lines (JSONL)**, also called **NDJSON (Newline Delimited JSON)**, stores **one valid JSON object per line**.

     {"emp_id":101,"emp_name":"John","department":"IT","salary":60000}
     {"emp_id":102,"emp_name":"Alice","department":"HR","salary":50000}
     {"emp_id":103,"emp_name":"Bob","department":"Finance","salary":70000}

- `Each line is a complete JSON object`.
- `There is no enclosing array ([])`.
- `Objects are not separated by commas`.

##### Multi-line JSON
- we set **multiline** option to **true** to read JSON records on file from **multiple lines**.
- By **default**, this option is set to **false**.
- is used when your JSON data is more complex and spans **multiple lines**.
- To correctly read **multi-line** JSON in Spark, you need to tell Spark that the **JSON objects** may span **multiple lines**.
- understanding these two formats is crucial because using the wrong method can lead to errors or incomplete data being loaded into Spark.

     [
       {"emp_id":101,"emp_name":"John","department":"IT","salary":60000},
       {"emp_id":102,"emp_name":"Alice","department":"HR","salary":50000},

**File Splitting Is Disabled:**
- When **multiline=True** is enabled, PySpark **cannot split** the file into **smaller chunks** to process in **parallel**.

##### 3) Explicit schema (Best practice)

- **Faster** than `inference`.
- **Prevents datatype / schema inference issues**.

     # Method 01
     spark.read.json('/Volumes/azureadb/pyspark/training/json/read_json/multiline.json', multiLine=True, schema=shema)

     # Method 02
     df = spark.read \
               .schema(schema) \
               .option("multiLine", "true") \
               .json("/Volumes/azureadb/pyspark/training/json/read_json/multiline.json")

```python
from pyspark.sql.types import StructField, StructType, StringType, IntegerType, LongType

schema = StructType([StructField('RecordNumber', IntegerType(), False),
                     StructField('State', StringType(), False),
                     StructField('ZipCodeType', StringType(), False),
                     StructField('Zipcode', LongType(), False),
                     StructField('country', StringType(), False),
                     StructField('description', StringType(), False),
                     StructField('input_timestamp', LongType(), False),
                     StructField('last_update_timestamp', LongType(), False),
                     StructField('source', StringType(), False),
                     StructField('user', StringType(), False)]
                    )

df_expl = spark.read.json('/Volumes/azureadb/pyspark/training/json/read_json/MultiLine/multiline.json', multiLine=True, schema=schema)
display(df_expl)

df_expl.printSchema()
```

##### 5) PERMISSIVE mode & explicit schema without `_corrupt_record`

```python
from pyspark.sql.types import StructField, StructType, IntegerType, StringType, LongType

# Define the main schema including the nested structure
Schema = StructType([StructField('source', StringType(), False),
                     StructField('description', StringType(), False),
                     StructField('input_timestamp', LongType(), False),
                     StructField('last_update_timestamp', LongType(), False),
                     StructField('country', StringType(), False),
                     StructField('user', StringType(), False),
                     StructField('State', StringType(), False),
                     StructField('RecordNumber', LongType(), False),
                     StructField('Zipcode', LongType(), False),
                     StructField('ZipCodeType', StringType(), False)]
                     )

df_multi_corrupt = spark.read \
    .format("json")\
    .schema(Schema)\
    .option('multiLine', True)\
    .option("mode", "PERMISSIVE")\
    .load("/Volumes/azureadb/pyspark/training/json/read_json/MultiLine/multiline_single_json_object_corrupt_record.json")

display(df_multi_corrupt)
```

```python
from pyspark.sql.types import StructField, StructType, IntegerType, StringType, LongType

# Define the main schema including the nested structure
Schema = StructType([StructField('source', StringType(), False),
                     StructField('description', StringType(), False),
                     StructField('input_timestamp', LongType(), False),
                     StructField('last_update_timestamp', LongType(), False),
                     StructField('country', StringType(), False),
                     StructField('user', StringType(), False),
                     StructField('State', StringType(), False),
                     StructField('RecordNumber', LongType(), False),
                     StructField('Zipcode', LongType(), False),
                     StructField('ZipCodeType', StringType(), False),
                     StructField("_corrupt_record", StringType(), True)]
                     )

df_multi_corrupt = spark.read \
    .format("json")\
    .schema(Schema)\
    .option('multiLine', True)\
    .option("mode", "PERMISSIVE")\
    .load("/Volumes/azureadb/pyspark/training/json/read_json/MultiLine/multiline_single_json_object_corrupt_record.json")

display(df_multi_corrupt)
```
