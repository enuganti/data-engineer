##### 3) Explicit schema (Best practice)

- **Faster** than `inference`.
- **Prevents datatype / schema inference issues**.

```python
from pyspark.sql.types import StructField, StructType, IntegerType, StringType, LongType

# Define custom schema for the nested structure
AddSchema_Address = StructType([StructField('country', StringType(), False),
                                StructField('user', StringType(), False),
                                StructField('Location', StringType(), False),
                                StructField('Zipcode', StringType(), False)]
                              )

AddSchema_domain = StructType([StructField('Level', StringType(), False),
                               StructField('technology', StringType(), False),
                               StructField('experience', StringType(), False),
                               StructField('Age', StringType(), False)]
                              )

# Define the main schema including the nested structure
Schema = StructType([StructField('S.No', IntegerType(), False),
                     StructField('First_Name', StringType(), False),
                     StructField('Last_Name', StringType(), False),
                     StructField('source', StringType(), False),
                     StructField('description', StringType(), False),
                     StructField('input_timestamp', LongType(), False),
                     StructField('last_update_timestamp', LongType(), False),
                     StructField('country', StringType(), False),
                     StructField('Address', AddSchema_Address),
                     StructField('domain', AddSchema_domain)]
                     )
```

```python
path = "/Volumes/azureadb/pyspark/training/json/read_json/SingleLine_Nested/singleline_nested_json.json"

# Load the JSON file using the custom schema
df_Snested = spark.read.format("json")\
                       .schema(Schema)\
                       .load(path)

display(df_Snested.limit(10))
df_Snested.printSchema()
```

##### 5) PERMISSIVE mode & explicit schema without `_corrupt_record`

```python
from pyspark.sql.types import StructField, StructType, IntegerType, StringType, LongType

# Define custom schema for the nested structure
AddSchema_Address = StructType([StructField('country', StringType(), False),
                                StructField('user', StringType(), False),
                                StructField('Location', StringType(), False),
                                StructField('Zipcode', StringType(), False)]
                              )

AddSchema_domain = StructType([StructField('Level', StringType(), False),
                               StructField('technology', StringType(), False),
                               StructField('experience', StringType(), False),
                               StructField('Age', StringType(), False)]
                              )

# Define the main schema including the nested structure
Schema = StructType([StructField('S.No', IntegerType(), False),
                     StructField('First_Name', StringType(), False),
                     StructField('Last_Name', StringType(), False),
                     StructField('source', StringType(), False),
                     StructField('description', StringType(), False),
                     StructField('input_timestamp', LongType(), False),
                     StructField('last_update_timestamp', LongType(), False),
                     StructField('country', StringType(), False),
                     StructField('Address', AddSchema_Address),
                     StructField('domain', AddSchema_domain)]
                     )
```

```python
df_multi_corrupt_wo = spark.read \
    .format("json")\
    .schema(Schema)\
    .option('multiLine', True)\
    .option("mode", "PERMISSIVE")\
    .load("/Volumes/azureadb/pyspark/training/json/read_json/SingleLine_Nested/singleline_nested_json_corrupt_record.json")

display(df_multi_corrupt_wo)
```

```python
from pyspark.sql.types import StructField, StructType, IntegerType, StringType, LongType

# Define custom schema for the nested structure
AddSchema_Address = StructType([StructField('country', StringType(), False),
                                StructField('user', StringType(), False),
                                StructField('Location', StringType(), False),
                                StructField('Zipcode', StringType(), False)]
                              )

AddSchema_domain = StructType([StructField('Level', StringType(), False),
                               StructField('technology', StringType(), False),
                               StructField('experience', StringType(), False),
                               StructField('Age', StringType(), False)]
                              )

# Define the main schema including the nested structure
Schema_corrupt = StructType([StructField('S.No', IntegerType(), False),
                             StructField('First_Name', StringType(), False),
                             StructField('Last_Name', StringType(), False),
                             StructField('source', StringType(), False),
                             StructField('description', StringType(), False),
                             StructField('input_timestamp', LongType(), False),
                             StructField('last_update_timestamp', LongType(), False),
                             StructField('country', StringType(), False),
                             StructField('Address', AddSchema_Address),
                             StructField('domain', AddSchema_domain),
                             StructField("_corrupt_record", StringType(), True)]
                            )
```

```python
df_multi_corrupt = spark.read \
    .format("json")\
    .schema(Schema_corrupt)\
    .option('multiLine', True)\
    .option("mode", "PERMISSIVE")\
    .load("/Volumes/azureadb/pyspark/training/json/read_json/SingleLine_Nested/singleline_nested_json_corrupt_record.json")

display(df_multi_corrupt)
```

```python
df_multi_corrupt_false = spark.read \
    .format("json")\
    .schema(Schema_corrupt)\
    .option('multiLine', False)\
    .option("mode", "PERMISSIVE")\
    .load("/Volumes/azureadb/pyspark/training/json/read_json/SingleLine_Nested/singleline_nested_json_corrupt_record.json")

display(df_multi_corrupt_false)
```
