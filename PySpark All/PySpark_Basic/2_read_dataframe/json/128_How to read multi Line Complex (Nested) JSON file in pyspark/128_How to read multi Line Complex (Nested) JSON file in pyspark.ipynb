##### 3) Explicit schema (Best practice)

- **Faster** than `inference`.
- **Prevents datatype / schema inference issues**.

```python
from pyspark.sql.types import StructField, StructType, IntegerType, StringType, LongType

# Define custom schema for the nested structure
AddSchema_Address = StructType([StructField('countryNM', StringType(), False),
                                StructField('userName', StringType(), False),
                                StructField('Location', StringType(), False),
                                StructField('pincode', StringType(), False)]
                              )

AddSchema_domain = StructType([StructField('Level', StringType(), False),
                               StructField('technology', StringType(), False),
                               StructField('experience', IntegerType(), False),
                               StructField('Age', IntegerType(), False)]
                              )

# Define the main schema including the nested structure
Schema = StructType([StructField('RecordNumber', IntegerType(), False),
                     StructField('source', StringType(), False),
                     StructField('description', StringType(), False),
                     StructField('input_timestamp', StringType(), False),
                     StructField('last_update_timestamp', StringType(), False),
                     StructField('country', StringType(), False),
                     StructField('user', StringType(), False),
                     StructField('State', StringType(), False),
                     StructField('Zipcode', LongType(), False),
                     StructField('ZipCodeType', StringType(), False),
                     StructField('Address', AddSchema_Address),
                     StructField('domain', AddSchema_domain)]
                     )
```

```python
path = "/Volumes/azureadb/pyspark/training/json/read_json/MultiLine_Nested/multiline_complex_nested.json"

# Load the JSON file using the custom schema
df_Mnested = spark.read.format("json")\
                       .option("multiline","true") \
                       .schema(Schema)\
                       .load(path)

display(df_Mnested)
df_Mnested.printSchema()
```

##### 5) PERMISSIVE mode & explicit schema without `_corrupt_record`

```python
from pyspark.sql.types import StructField, StructType, IntegerType, StringType, LongType

# Define custom schema for the nested structure
AddSchema_Address = StructType([StructField('countryNM', StringType(), False),
                                StructField('userName', StringType(), False),
                                StructField('Location', StringType(), False),
                                StructField('pincode', StringType(), False)]
                              )

AddSchema_domain = StructType([StructField('Level', StringType(), False),
                               StructField('technology', StringType(), False),
                               StructField('experience', StringType(), False),
                               StructField('Age', StringType(), False)]
                              )

# Define the main schema including the nested structure
Schema_wo_corpt = StructType([StructField('RecordNumber', IntegerType(), False),
                              StructField('source', StringType(), False),
                              StructField('description', StringType(), False),
                              StructField('input_timestamp', StringType(), False),
                              StructField('last_update_timestamp', StringType(), False),
                              StructField('country', StringType(), False),
                              StructField('user', StringType(), False),
                              StructField('State', StringType(), False),
                              StructField('Zipcode', LongType(), False),
                              StructField('ZipCodeType', StringType(), False),
                              StructField('Address', AddSchema_Address),
                              StructField('domain', AddSchema_domain)]
                            )
```

```python
df_multi_corrupt_wo = spark.read \
    .format("json")\
    .schema(Schema_wo_corpt)\
    .option('multiLine', True)\
    .option("mode", "PERMISSIVE")\
    .load("/Volumes/azureadb/pyspark/training/json/read_json/MultiLine_Nested/multiline_complex_nested_corrupt_record.json")

display(df_multi_corrupt_wo)
df_multi_corrupt_wo.printSchema()
```

```python
from pyspark.sql.types import StructField, StructType, IntegerType, StringType, LongType

# Define custom schema for the nested structure
AddSchema_Address = StructType([StructField('countryNM', StringType(), False),
                                StructField('userName', StringType(), False),
                                StructField('Location', StringType(), False),
                                StructField('pincode', StringType(), False)]
                              )

AddSchema_domain = StructType([StructField('Level', StringType(), False),
                               StructField('technology', StringType(), False),
                               StructField('experience', StringType(), False),
                               StructField('Age', StringType(), False)]
                              )

# Define the main schema including the nested structure
Schema_w_corpt = StructType([StructField('RecordNumber', IntegerType(), False),
                             StructField('source', StringType(), False),
                             StructField('description', StringType(), False),
                             StructField('input_timestamp', StringType(), False),
                             StructField('last_update_timestamp', StringType(), False),
                             StructField('country', StringType(), False),
                             StructField('user', StringType(), False),
                             StructField('State', StringType(), False),
                             StructField('Zipcode', LongType(), False),
                             StructField('ZipCodeType', StringType(), False),
                             StructField('Address', AddSchema_Address),
                             StructField('domain', AddSchema_domain),
                             StructField("_corrupt_record", StringType(), True)]
                           )
```

```python
df_multi_corrupt_false = spark.read \
    .format("json")\
    .schema(Schema_w_corpt)\
    .option('multiLine', True)\
    .option("mode", "PERMISSIVE")\
    .load("/Volumes/azureadb/pyspark/training/json/read_json/MultiLine_Nested/multiline_complex_nested_corrupt_record.json")

display(df_multi_corrupt_false)
```
