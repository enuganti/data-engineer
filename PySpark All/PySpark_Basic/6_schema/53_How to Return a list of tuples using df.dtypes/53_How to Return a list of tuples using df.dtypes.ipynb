##### How to Return a `list of tuples (columnName & dataType)` using `df.dtypes`?

##### df.dtypes

- Returns a **list of tuples**, where **each tuple** contains the **column name** and its corresponding **data type**.
  - `tuple` -> `(column_name, data_type)`

     # df.dtypes
     [('start_date', 'date'),
      ('product_url', 'string'),
      ('category', 'string'),
      ('default_group', 'string')]

|     Method       |     Output Type          |      Best Use Case                              |
|------------------|--------------------------|-------------------------------------------------|
| df.printSchema() | None (Prints to console) | Quick visual inspection of tree structure       |
| df.schema        | StructType               | Programmatic schema manipulation and validation |
| df.columns       | List[str]                | Getting just the names of the columns           |
| df.dtypes        | List[Tuple[str, str]]    | Quick mapping of names to simple type strings.  |

**Key Features of df.dtypes**

**Format:**
- Returns a **list of strings and tuples**, unlike the **df.schema** object which returns complex DataType objects.

**Direct Access:**
- Useful for **checking if a specific column exists** and what its **data type** is **without iterating** through the **schema**.

**Efficiency:**
- This is a **metadata** operation and **does not trigger** a **Spark job or scan the data**.

**Filtering:**
- You can easily **filter** for specific **data types** using a **dictionary comprehension**:
  - `string_cols = [k for k, v in dtype_dict.items() if v == 'string']`

```python
from pyspark.sql import functions as f
```

```python
df = df.withColumn("load_date", f.current_date()) \
       .withColumn("load_time", f.current_timestamp()) \
       .withColumn("current_user", f.current_user()) \
       .withColumn("reach", f.lit(10.55))

display(df.limit(5))
```

##### 1) How to Verify the data type of the Dataframe?

```python
# Return data type of all columns
# Returns a list of (column_name, data_type) tuples
print(f'Data types of all the columns is : {df.dtypes}')
```

```python
df.dtypes
```

```python
data_type = dict(df.dtypes)
data_type
```

##### 2) How to Verify the specific column data type of the Dataframe using dict()?

```python
# finding data type of the 'source_target' column using dtype function
data_type = dict(df.dtypes)['source_target']

# printing
print(f'Data type of source_target is : {data_type}')

# visualizing the dataframe
display(df.limit(5))
```

```python
print(dict(df.dtypes)['source_target'])
print(dict(df.dtypes)['sessions'])
print(dict(df.dtypes)['load_time'])
print(dict(df.dtypes)['reach'])
```

##### 3) How to Verify the column type of the Dataframe using for loop?

```python
print("Datatype of the columns with column names are:")

# finding datatype of all column with column name using for loop
for col in df.dtypes:
  
  # printing the column and datatype of that column
  print(col[0], ",", col[1])

# visualizing the dataframe
display(df.limit(5))
```

**4) How to return the first column's type if there are multiple columns with the same name?**

```python
def get_dtype(df,colname):
    return [dtype for name, dtype in df.dtypes if name == colname][0]

print(get_dtype(df,'start_date'))
print(get_dtype(df,'product_url'))
print(get_dtype(df,'sessions'))
print(get_dtype(df,'load_time'))
```

##### 5) How to filter Columns by Type?

```python
numeric_cols = [name for name, dtype in df.dtypes if dtype in ['int', 'double', 'float']]
print(numeric_cols)

df_int = df.select(*numeric_cols)
display(df_int.limit(5))
```

```python
str_cols = [name for name, dtype in df.dtypes if dtype == 'string']
print(str_cols)

df_str = df.select(*str_cols)
display(df_str.limit(5))
```
