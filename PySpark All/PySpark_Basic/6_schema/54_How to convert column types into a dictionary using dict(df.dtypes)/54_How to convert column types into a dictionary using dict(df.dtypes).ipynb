##### How to convert column types into a dictionary using dict(df.dtypes)?

##### df.dtypes

- Returns a **list of tuples**, where **each tuple** contains the **column name** and its corresponding **data type**.
  - `tuple` -> `(column_name, data_type)`

##### dict(df.dtypes)

- `dict()` constructor transforms this `list into a dictionary` where **column names are keys** and **data types are values**.

     # df.dtypes
     [('start_date', 'date'),
      ('product_url', 'string'),
      ('category', 'string'),
      ('default_group', 'string')]

     # dict(df.dtypes)
     {'start_date': 'date',
      'product_url': 'string',
      'category': 'string',
      'default_group': 'string'}

     # Convert to dictionary
     dtype_dict = dict(df.dtypes)

     # Access a specific column's type
     category_type = dtype_dict['category']
     print(category_type)  # Output: 'string'

```python
df = spark.read.csv("/Volumes/azureadb/pyspark/unionby/company_level.csv", header=True, inferSchema=True)
display(df.limit(5))

df.printSchema()
```

```python
from pyspark.sql import functions as f
```

```python
df = df.withColumn("load_date", f.current_date()) \
       .withColumn("load_time", f.current_timestamp()) \
       .withColumn("current_user", f.current_user()) \
       .withColumn("rating", f.lit(15.89))

display(df.limit(5))

df.printSchema()
```

##### 5) Find numeric columns

##### 7) Rename only columns of a specific datatype

```python
dtype_dict = dict(df.dtypes)

df_str = df
for col, type in dtype_dict.items():
    if type == "string":
        df_str = df_str.withColumnRenamed(col, col.upper())

display(df_str.limit(5))
```

```python
dtype_dict = dict(df.dtypes)

df_int = df
for col, type in dtype_dict.items():
    if type in ("int", "integer", "double", "float", "long", "bigint", "smallint"):
        df_int = df_int.withColumnRenamed(col, col.upper())

display(df_int.limit(5))
```

```python
dtype_dict = dict(df.dtypes)

df_all = df
for col, type in dtype_dict.items():
        df_all = df_all.withColumnRenamed(col, col.upper())

display(df_all.limit(5))
```

##### 8) Validate column types before transformation

```python
dtype_dict = dict(df.dtypes)

if dtype_dict.get("status_name") == "string":
    print("status_name column is string")
else:
    raise ValueError("status_name column must be string")
```

```python
dtype_dict = dict(df.dtypes)

if dtype_dict.get("start_date") == "string":
    print("start_date column is string")
else:
    raise ValueError("start_date column must be date")
```

##### 9) Use it for dynamic casting

```python
import pyspark.sql.functions as F

dtype_dict = dict(df.dtypes)

df_long = df
for col, typ in dtype_dict.items():
    if typ == "int":
        df_long = df_long.withColumn(col, F.col(col).cast("long"))

display(df_long.limit(5))

df_long.printSchema()
```

```python
import pyspark.sql.functions as F

dtype_dict = dict(df.dtypes)

df_dt = df
for col, typ in dtype_dict.items():
    if typ == "date":
        df_dt = df_dt.withColumn(col, F.col(col).cast("timestamp"))

display(df_dt.limit(5))

df_dt.printSchema()
```
