##### 1) How to Create DataFrame with `empty strings` (list of tuples)?

```python
data = [
    (1, "", "IT", 5600, "Chennai"),
    (2, "Ravi", "", 6250, ""),
    (3, "", "", 7234, "Bangalore"),
    (4, "Sankar", "Admin", 5600, "Hyderabad"),
    (5, "Senthil", "", 9872, ""),
    (6, "", "Sales", 1235, "Amaravati"),
    (7, "", "Marketing", 9567, "Delhi"),
    (8, "Smriti", "", 3478, ""),
    (9, "", "Finance", 8734, "Noida"),
]

columns = ["emp_id", "emp_name", "department", "salary", "city"]

df_empty = spark.createDataFrame(data, columns)
display(df_empty)
```

##### 2) How to Create DataFrame using `explicit schema`?

- Ensures `empty string` is preserved.
- Recommended for `production`.

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType
```

```python
schema = StructType([
    StructField("emp_id", IntegerType(), True),
    StructField("emp_name", StringType(), True),
    StructField("department", StringType(), True),
    StructField("salary", IntegerType(), True),
    StructField("city", StringType(), True)
])

data = [
    (1, "", "IT", 5600, "Chennai"),
    (2, "Ravi", "", 6250, ""),
    (3, "", "", 7234, "Bangalore"),
    (4, "Sankar", "Admin", 5600, "Hyderabad"),
    (5, "Senthil", "", 9872, ""),
    (6, "", "Sales", 1235, "Amaravati"),
    (7, "", "Marketing", 9567, "Delhi"),
    (8, "Smriti", "", 3478, ""),
    (9, "", "Finance", 8734, "Noida"),
]

df_empty_schema = spark.createDataFrame(data, schema)
display(df_empty_schema)
```

##### 3) How to Create table with `empty strings` using `Spark SQL`?

```python
spark.sql("""
    CREATE OR REPLACE TEMP VIEW employees AS
    SELECT 1 AS emp_id, '' AS emp_name, 'IT' AS department, 5600 AS Salary, 'Chennai' AS city
    UNION ALL
    SELECT 2, 'Ravi', '', 6250, ''
    UNION ALL
    SELECT 3, '', '', 7234, 'Bangalore'
    UNION ALL
    SELECT 4, 'Sankar', 'Admin', 5600, ''
    UNION ALL
    SELECT 5, 'Senthil', '', 9872, ''
    UNION ALL
    SELECT 6, '', 'Marketing', 5780, 'Hyderabad'
""")

spark.sql("SELECT * FROM employees").display()
```

##### 4) How to Create `empty-string` columns using `lit("")`?

```python
from pyspark.sql.functions import lit

df_lit = spark.range(1, 8) \
    .withColumn("emp_name", lit("")) \
    .withColumn("department", lit("")) \
    .withColumn("salary", lit(25000))

display(df_lit)
```

##### 5) How to Create `empty DataFrame` with `string columns`?

- **No rows**
- **Columns exist**

```python
columns = ["emp_id", "emp_name", "department"]

empty_df_list_wo = spark.createDataFrame([], columns)
display(empty_df_list_wo)
```

```python
columns = ["emp_id", "emp_name", "department"]

empty_df_list = spark.createDataFrame([(1, "Raghu", "IT")], columns)
display(empty_df_list)
```

```python
schema = StructType([
    StructField("emp_id", IntegerType(), True),
    StructField("emp_name", StringType(), True),
    StructField("department", StringType(), True)
])

empty_df = spark.createDataFrame([], schema)
display(empty_df)
```

##### 6) How to Create managed table with empty strings?

```python
df_lit.write.mode("overwrite").option("mergeSchema", "true").saveAsTable("employee_table")
```
