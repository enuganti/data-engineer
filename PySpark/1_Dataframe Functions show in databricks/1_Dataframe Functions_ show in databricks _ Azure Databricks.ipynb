By default, PySpark displays only the **first 20 rows** and truncates each **column to 20 characters**. To change this, you can adjust the following configurations:

- **How to show Full Data:** Use the show() method with the truncate parameter set to False:
     
       df.show(truncate=False)

- **How to display More than 20 Rows:** If you want to display more than 20 rows, specify the number of rows as the first parameter in the show() method:
     
       df.show(n=100, truncate=False)

```
 help(df.show)
```

```
df = spark.read.csv("dbfs:/FileStore/tables/titanic.csv", header=True, inferSchema=True)
df.show()
```

```
# prints 20 records by default
df.show()
```

```
# Both gives same results
# df.show()
df.show(truncate=True)    
```

```
# show df created top 10 rows
df.show(10)
```

```
 # visualizing full content of the Dataframe by setting truncate to False
 # Both gives same results
 df.show(truncate=False)
 #df.show(truncate=0)
```

```
#df.show(truncate=1)
#df.show(truncate=5)
df.show(truncate=45)
```

```
# Show top 2 rows and full column contents (PySpark)
df.show(2, truncate=False)
```

```
# Shows top 2 rows and only 25 characters of each column (PySpark)
df.show(2, truncate=25)
```

```
# get total row count and pass it as argument to show
df.show(df.count())
```

```
df.show(df.count(), truncate=False)
```

```
# Shows rows vertically (one line per column value) (PySpark)
df.show(vertical=True)
```

```
# Both gives same results
df.show(n=3, truncate=25, vertical=True)
#df.show(3, 25, True)
```
