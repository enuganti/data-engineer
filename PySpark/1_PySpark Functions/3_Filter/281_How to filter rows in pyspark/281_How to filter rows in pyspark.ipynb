##### Filter

- PySpark **filter()** function is used to **filter the rows** from **DataFrame** based on the given **condition or SQL expression**.
- if you are coming from an `SQL background`, you can also use **where()** clause **instead of the filter().**
- **both these functions** operate exactly the **same**.

**Syntax:**

     df.filter(#condition)
     df.where(#condition)

- **df:** The PySpark DataFrame
- **Condition:** The Filter condition which we want to Implement on.

| Operation    | Example                                        |
| ------------ | ---------------------------------------------- |
| Greater than | `col("age") > 18`                              |
| Equal        | `col("city") == "Bangalore"`                   |
| Between      | `col("dt").between("2025-01-01","2025-12-31")` |
| NULL         | `.isNull()`                                    |
| IN           | `.isin()`                                      |
| Multiple     | `&`, `\|`, `~`                                 |

**Common Filtering Methods**

|  Method       | 	 Description	                                                      |    Example                                        |
|---------------|-----------------------------------------------------------------------|---------------------------------------------------|
|  isin()       | Filters for rows where a column's value is in a given list of values. | df.filter(df["department"].isin("HR", "Finance")) |
|  isNull()     | Selects rows where a column has a NULL value.                         | df.filter(df["age"].isNull())                     |
|  isNotNull()  | Selects rows where a column does not have a NULL value.               | df.filter(df["age"].isNotNull())                  |
|  startswith() | Filters string columns that start with a specific substring.          | df.filter(df["name"].startswith("A"))             |
|  endswith()   | Filters string columns that end with a specific substring.            | df.filter(df["name"].endswith("son"))             |
|  like()       | Filters using simple SQL wildcards (e.g., _ for single character, % for multiple). | df.filter(df["name"].like("A%"))     |
|  rlike()      | Filters using a regular expression for complex pattern matching.      | df.filter(df["name"].rlike("son"))                |
|  contains()   | Checks if a column's string value contains a specified substring.     | df.filter(df["location"].contains("google.com"))  |

**---> Filtering with PySpark**

     # Filter using a column condition
     filtered_df = df.filter(df["age"] > 30)

     # Filter using the alias 'where' (same result)
     filtered_df = df.where(col("age") > 30)

**---> Filtering with SQL Queries**
- For users more comfortable with **SQL syntax**, you can register a DataFrame as a **temporary view and run SQL queries**.

##### Syntax
     # Register the DataFrame as a temporary table 
     df.createOrReplaceTempView("employees")

     # Filter using an SQL expression string
     filtered_df = df.filter("age > 30")

     # SQL query to filter rows with "with milk"
     filtered_df = spark.sql("""SELECT * FROM transactions
                                WHERE Description LIKE '%with milk%'
                             """)

     # Use SQL query for filtering
     filtered_df = spark.sql("SELECT * FROM employees WHERE age > 30")

##### a) Basic Filtering: Using filter() and where() Methods
- `Filter by Column instances`.

      # PySpark
      df.filter(df.age > 3).display()
      df.where(df.age == 2).display()

      # SQL
      df.filter("age > 3").display()
      df.where("age = 2").display()

##### b) Combining Conditions with Logical Operators
- You can `filter` based on **multiple conditions** using **logical operators** like **& (AND), | (OR), and ~ (NOT)**.
- These **logical operators** are particularly useful when filtering on **multiple columns**.

|  Operator   |  	Description    |
|-------------|------------------|
|     &	      |  AND condition   |
|    \|       |  OR Condition    |
|     ~	      |  NOT condition   |

        # Single filter Condition
        df.filter(df['Salary']<=20000).show()
        df.filter(~(df['Salary']<=20000)).show()

        # Multiple filter Conditions
        df.filter((df['Salary']<=20000) & (df['Salary']>=15000)).show()
        df.filter((df['Salary']<=20000) | (df['Salary']>=15000)).show()

- Always **wrap conditions** in **parentheses**.
- filter() does **row-level filtering**.
- filter() does **not modify the original DataFrame**.

##### Filter by multiple conditions.

     # PySpark
     df.filter((df.age > 3) & (df.subject == "Physics")).display()
     df.filter((df.age == 2) | (df.subject == "Chemistry")).display()

     # SQL
     df.filter("age > 3 AND name = 'Bob'").display()

##### Common Mistakes

     ❌ Wrong:
     df.filter(col("age") > 18 and col("salary") > 50000)

     ✔ Correct:
     df.filter((col("age") > 18) & (col("salary") > 50000))

##### c) Filter using the Column.isin() function
- The isin() method is ideal when filtering based on a **set of values**, similar to **SQL’s IN operator**.
- you can efficiently `filter` **multiple values** within a **single column**.
- This is useful for checking membership in a specific **list or set**.

      # PySpark
      df.filter(col("city").isin("Bangalore", "Chennai", "Hyderabad"))

      df.filter(col("subject").isin(["Bangalore", "Chennai", "Hyderabad"]))

      df.filter(~col("name").isin(["Bangalore", "Chennai", "Hyderabad"]))

      # Equivalent to SQL:
      WHERE city IN (...)

##### d) Filter using the Column.between() function

     # PySpark
     df.filter(df.age.between(2, 5))

     df.filter(col("report_dt").between("2025-07-01", "2025-12-31"))

     # Equivalent to:
     df.filter(
        (col("report_dt") >= "2025-07-01") &
        (col("report_dt") <= "2025-12-31")
    )

##### e) Filtering with startswith()
- The startswith() function is used to `filter rows` where the value in a column **starts** with a **specific substring**.
- This is similar to using **LIKE 'abc%' in SQL**.
- This would `keep rows` where the `name column begins` with the letter **"A"**.
- This function is **case-sensitive**, so it `won’t match` **lowercase 'a' unless specified**.

     # Filtering rows where the name starts with 'A'
     filtered_df = df.filter(df["name"].startswith("A"))

##### f) Filtering with endswith()
- Similarly, the `endswith()` function allows `filtering rows` based on whether a column value **ends with a specified substring**.
- This works like **SQL’s LIKE '%xyz'**.
- This will `keep rows` where the **name column ends** with **"son", such as "Anderson" or "Jackson"**.

     # Filtering rows where the name ends with 'son'
     filtered_df = df.filter(df["name"].endswith("son"))

##### g) Null Handling: isNull() and isNotNull()
- You can use **isNull() and isNotNull()** to specifically `filter rows` based on the **presence or absence** of `null values`.

     # Filter rows where age is NULL
     filtered_df = df.filter(df["age"].isNull())

     # Filter rows where age is NOT NULL
     filtered_df = df.filter(df["age"].isNotNull())

##### h) Filtering with like() and rlike() for Pattern Matching
- For `pattern matching`, PySpark provides the **like() and rlike()** functions, which work similarly to **SQL’s LIKE and REGEXP operators**.

**Basic Pattern Matching with like():**
- This is helpful for **simple wildcard filtering**, such as `filtering names that start with a particular letter`.

      # Names starting with 'A'
      filtered_df = df.filter(df["name"].like("A%"))

**Regex Pattern Matching with rlike():**
- Use rlike() for more **complex patterns**, as it supports **regular expressions**.

      # Names containing 'son'
      filtered_df = df.filter(df["name"].rlike("son"))
