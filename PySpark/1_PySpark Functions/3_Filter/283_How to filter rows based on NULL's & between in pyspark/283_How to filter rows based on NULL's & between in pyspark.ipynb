```python
import pyspark.sql.functions as F
from pyspark.sql.functions import col, to_date
```

```python
data = [
    (1, "Suresh", 25, "IT", 50000),
    (2, "Ravi", 30, "HR", 40000),
    (3, "Anita", 28, "IT", None),
    (4, "Kiran", 35, "Finance", 70000),
    (5, "Meena", None, "", 45000),
    (6, "Rajesh", 32, "IT", 55000),
    (7, "", 29, "IT", 50000),
    (8, "Sandeep", 31, "Admin", 55000),
    (9, "Rakesh", 33, "IT", 50000),
    (10, "Suman", 34, "Accounts", None)
]

columns = ["id", "name", "age", "dept", "salary"]

df = spark.createDataFrame(data, columns)
display(df)
```

     df.filter("state is NULL").display()

**a) Filter Null Values:**

- **Null values** can only be queried using **isNull()** attribute of col function. Rows were fetched where manager_id was null in the below example.

```python
# Rows where age is NULL
df.filter(col("age").isNull()).display()
```

**b) Filter not Null Values:**

- **isNotNull** attribute of col function can be used to filter out `null values`.

```python
# Rows where age is NOT NULL
df.filter(col("age").isNotNull()).display()
```

##### 2) Filter using Between

- Between attribute of col function can be used to `filter` data from a column based on **lower and upper range**.

```python
df_fltr = spark.read.csv("/Volumes/azureadb/pyspark/filter/filter_records.csv", inferSchema=True, header=True)
display(df_fltr.limit(30))
print("No of Records: ", df_fltr.count())
df_fltr.printSchema()
```

```python
df_fltr.select("report_dt").distinct().orderBy("report_dt").display()
```

```python
df_fltr_final = df_fltr.filter(col("report_dt").between("2024-01-01", "2024-06-30"))
display(df_fltr_final.limit(30))
print("No of filter records: ", df_fltr_final.count())
```

```python
df_fltr_final.select("report_dt").distinct().orderBy("report_dt").display()
```
