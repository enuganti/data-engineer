```python
data = [
    (1, "Suresh", 25, "IT", 50000),
    (2, "Ravi", 30, "HR", 40000),
    (3, "Anita", 28, "IT", None),
    (4, "Kiran", 35, "Finance", 70000),
    (5, "Meena", None, "", 45000),
    (6, "Rajesh", 32, "IT", 55000),
    (7, "", 29, "IT", 50000),
    (8, "Sandeep", 31, "Admin", 55000),
    (9, "Rakesh", 33, "IT", 50000),
    (10, "Suman", 34, "Accounts", None)
]

columns = ["id", "name", "age", "dept", "salary"]

df = spark.createDataFrame(data, columns)
display(df)
```

##### 1) DataFrame `filter()` by `Column Condition`
- Filter Using a `Single Condition`.
- Employees dept `IT`.

**a) equal condition**

```python
# Using equals condition
df.filter(df.dept == "IT").display()
```

##### b) not equals condition

     # Both gives same output
     df.filter(df.dept != "IT").display()
                          (or)
     df.filter(~(df.dept == "IT")).display()

```python
# not equals condition
df.filter(df.dept == "IT").display()
df.filter(~(df.dept == "IT")).display()
```

##### 2) DataFrame `filter()` with `SQL Expression`

```python
# Using SQL Expression
df.filter("dept == 'IT'").display()
```

```python
# For not equal
df.filter("dept != 'IT'").display()
df.filter("dept <> 'IT'").display()
```

```python
df.filter("salary >= 50000 AND dept = 'IT'").display()
```

##### 3) Filter with Multiple Conditions (Multicolumn filters)
- We can use **conditional operators** for filtering with conditions. We would be using **circular brackets** in case of filtering with **two conditions**. 

  - **~** is **not**
  - **&** is **and**
  - **|** is **or** 

        df.filter((df['Close'] <200) & ~(df['Open']>200)).show()

**Multicolumn filters:**

- Multiple columns can be used to filter data in dataframe.
- `Pipe(|)` can be used `between conditions` to perform `OR` operation as in SQL joins and ampersand(`&`) can be used between conditions to perform `AND` operation as in SQL joins.

```python
# Filter multiple condition
# IT employees with salary > 50,000
df.filter((df.dept == "IT") & (df.salary > 50000)).display()
```

```python
# HR OR Finance department
df.filter((df.dept == "HR") | (df.dept == "Finance")).display()
```

```python
data = [
    (7369, "Sunitha", 800, 200),
    (7499, "Swaroop", 1600, 300),
    (7521, "Varshini", 1250, 300),
    (7566, "Joseph", 2975, 200),
    (7654, "Mythili", 1250, 300),
    (7698, "Balaji", 2850, 300),
    (7782, "Sanjay", 2450, 100),
    (7788, "Senthil", 3000, 200),
    (7839, "Karthik", 5000, 100),
    (7844, "Tharun", 1500, 300),
    (7876, "Swaroop", 1100, 200),
    (7900, "Jagadish", 950, 300),
    (7902, "Sanjay", 3000, 200)
]

columns = ["emp_id", "emp_name", "salary", "dept_no"]

df1 = spark.createDataFrame(data, columns)
display(df1)
```

```python
# This will return rows where emp_name is either "Sanjay" or "Swaroop"
df1.filter((col("emp_name")=="Sanjay")|(col("emp_name")=="Swaroop")).display()
```

```python
# This will return rows where either emp_name is Sanjay or people having salary greater than 2000
df1.filter((col("emp_name")=="Sanjay")|(col("salary")>3000)).display()
```

```python
# This will return rows where people are part of dept_no 200 and have salary greater than 2000
df1.filter((col("dept_no")==200)&(col("salary")>2000)).display()
```

```python
# This will return rows where people are part of dept_no 200 and have salary greater than 2000 or emp_name is "Jagadish"
df1.filter((((col("dept_no")==200)&(col("salary")>2000))|(col("emp_name")=="Jagadish"))).display()
```
