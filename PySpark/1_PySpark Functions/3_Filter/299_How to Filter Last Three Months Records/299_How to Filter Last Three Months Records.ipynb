##### How to filter Last 7 Days Data?

```python
from pyspark.sql.functions import date_add, date_sub, add_months, col, current_date
```

```python
data = [
    (1, "A", "2026-03-10",),
    (2, "B", "2026-03-11",),
    (3, "C", "2026-03-12",),
    (4, "D", "2026-03-13",),
    (5, "E", "2026-03-14",),
    (6, "F", "2026-03-15",),
    (7, "G", "2026-03-16",),
    (8, "H", "2026-03-17",),
    (9, "I", "2026-03-18",),
    (10, "J", "2026-03-05",),
    (11, "K", "2026-02-28",),
    (12, "L", "2026-02-17",),
    (13, "M", "2026-02-08",)
]

df_fltr = spark.createDataFrame(data, ["id", "name", "load_date"]) \
           .withColumn("load_date", col("load_date").cast("date"))
display(df_fltr)
```

##### 1) Using current_date()

```python
df_last_7 = df_fltr.filter(
    col("load_date") >= date_add(current_date(), -7)
).display()
```

```python
from pyspark.sql.functions import current_date, date_sub

df_last_7 = df_fltr.filter(
    col("load_date") >= date_sub(current_date(), 7)
).display()
```

```python
df_last_7 = df_fltr.filter(
    col("load_date") >= add_months(current_date(), -1)
).display()
```

##### 2) Using current_timestamp()

```python
from pyspark.sql.functions import current_timestamp, expr

df_last_7 = df_fltr.filter(
    col("load_date") >= expr("current_timestamp() - INTERVAL 7 DAYS")
).display()
```
