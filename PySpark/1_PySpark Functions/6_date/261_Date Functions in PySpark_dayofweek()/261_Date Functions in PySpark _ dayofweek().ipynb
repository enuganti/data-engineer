**dayofweek():**
- is used to `extract the day of the week` from a `date or timestamp` column as an `integer`.
- `Returns day of week (1 = Sunday, 7 = Saturday)`
- `starts from Sunday = 1 (Not Monday!)`

**Syntax**

     dayofweek(expr)

**Arguments:**
- **expr:** A `DATE, TIMESTAMP, or TIMESTAMP_NTZ` expression.

**Returns:**
- An `INTEGER` where **1 = Sunday, and 7 = Saturday**.

##### Convert Day Number to Day Name

- `Since dayofweek() returns numbers`:

| Number | Day       |
| ------ | --------- |
| 1      | Sunday    |
| 2      | Monday    |
| 3      | Tuesday   |
| 4      | Wednesday |
| 5      | Thursday  |
| 6      | Friday    |
| 7      | Saturday  |

```python
import pyspark.sql.functions as F
from pyspark.sql.functions import col, date_format, dayofweek
```

```python
# define data
data = [['2023-04-02', 20],
        ['2023-04-08', 22],
        ['2023-04-15', 14],
        ['2023-04-17', 12],
        ['2023-05-21', 15],
        ['2023-05-23', 30],
        ['2023-10-26', 45],
        ['2023-10-28', 32],
        ['2023-10-29', 47]]
  
# define column names
columns = ['date', 'sales']
  
# create dataframe using data and column names
df_dow = spark.createDataFrame(data, columns) \
        .withColumn('date', col('date').cast('date')) \
        .withColumn('sales', col('sales').cast('integer'))

display(df_dow)
df_dow.printSchema()
```

##### 1) Get Day of Week as Number (Sunday =1)
- We can use the following syntax to get the **day of the week** as a number **between 1 and 7**, assuming **Sunday** is the **start of the week**.

```python
# Add new column that displays day of week as number
df_dow_sun = df_dow \
  .withColumn('day_name', F.date_format('date', 'EEEE')) \
  .withColumn('day_of_week', F.dayofweek('date'))

display(df_dow_sun)
```

##### 2) Get Day of Week as Number (Monday=1)
- To get the **day of the week** as a number **between 1 and 7**, assuming **Monday** is the **start of the week**.

```python
# Add new column that displays day of week as number
df_dow_mon = df_dow \
  .withColumn('day_name', F.date_format('date', 'EEEE')) \
  .withColumn('day_of_week', ((F.dayofweek('date')+5)%7)+1)

display(df_dow_mon)
```

##### 3) Get Day of Week as Abbreviated Name
- To get the **short name** of weekday we can use **E/EE/EEE** and if you want the **full name** of then to need to give `more than 3Es` like **EEEE**.

```python
# Add new column that displays day of week as abbreviated name
df_dow_abbr = df_dow.withColumn('day_of_week', F.date_format('date', 'E'))
display(df_dow_abbr)
```

##### 4) Get Day of Week as Full Name

```python
# Add new column that displays day of week as full name
df_dow_abbr_nm = df_dow.withColumn('day_of_week', F.date_format('date', 'EEEE'))
display(df_dow_abbr_nm)
```
