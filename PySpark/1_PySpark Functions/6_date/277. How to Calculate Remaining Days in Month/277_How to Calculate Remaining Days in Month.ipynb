##### How to Calculate Remaining Days in a Month?

**Common Use Cases**
- `Used in billing, reporting, and time-based analytics to drive decisions before month-end.`.

- `Subscription Tracking -> OTT platforms (Netflix, etc.)`
  - `To determine how many days are left in a billing cycle`.
  - `Interest calculation based on remaining days -> (Loan / EMI Calculations)`.
  - `Remaining days for sprint planning -> (Project Deadlines / Sprint Planning)`.

- `To calculate prorated charges`
  - `User subscribes on 20th → charged only for remaining days`.

##### Recommended Method: datediff and last_day

     Remaining days = last day of month − current date

**Step 1: Get the Last Day of month**
- Use `last_day(col)` to return the `final day of the month` for any given `date column`.

**Step 2: Calculate the Difference**
- Use `datediff(end, start)` to get the `number of days between the last day and your reference date`.

```python
from pyspark.sql.functions import last_day, datediff
```

```python
%sql
CREATE OR REPLACE TABLE subscriptions (
    customer_id INT,
    customer_name STRING,
    start_date DATE
);

INSERT INTO subscriptions VALUES
(1, 'Akash', '2025-03-05'),
(2, 'Ritesh', '2025-03-10'),
(3, 'Dravid', '2025-03-20'),
(4, 'Pushpa', '2025-02-15'),
(5, 'Sophia', '2025-04-01'),
(6, 'Ravi', '2025-04-10'),
(7, 'Priya', '2025-04-20'),
(8, 'Rajesh', '2025-05-01'),
(9, 'Priyanka', '2025-05-10'),
(10, 'Raj', '2025-05-20');

SELECT * FROM subscriptions;
```

```python
df_subscr = spark.table("subscriptions")

df_subscr_rp = df_subscr \
    .withColumn("last_day_of_month", last_day("start_date")) \
    .withColumn("days_remaining", datediff(last_day("start_date"), "start_date"))

display(df_subscr_rp)
```
