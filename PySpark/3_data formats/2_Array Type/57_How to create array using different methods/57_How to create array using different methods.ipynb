##### Array

- An `array` column must contain `elements` of the **same data type**.
- You `cannot` have a `single array` with **mixed types** like **string + int + boolean**.
  - `Mixed data types NOT allowed`.
  - `Arrays must be same data type`.
- If you need **mixed types** → use **STRUCT**.

     # NOT allowed (mixed types in one array)
     ["Spark", 10, True, 3.14]   ❌

     # Spark will throw => Array element type mismatch

| Use case                   | Method                    |
| -------------------------- | ------------------------- |
| Create array from literals | `array(lit(), lit())`     |
| From columns               | `array(col1, col2)`       |
| From source data           | `spark.createDataFrame()` |
| Different data types       | `struct()`                |

**Topics Covered**
- `Array of strings`
- `Array of integers`
- `Array of booleans`
- `Array of doubles`
- `Create array from existing columns`
- `Create array while creating DataFrame`
- `Use STRUCT instead of ARRAY`
- `Array of STRUCTs`

```python
from pyspark.sql.functions import array, lit
from pyspark.sql.types import ArrayType, IntegerType, StringType, BooleanType, DoubleType, StructType, StructField
```

```python
spark.range(6).display()
```

##### 1) Array of strings
- `Create array using array() + lit()`

```python
df_str = spark.range(6).select("id",
    array(
        lit("Spark"),
        lit("Python"),
        lit("SQL")
    ).alias("skills")
)

df_str.printSchema()
display(df_str)
```

```python
from pyspark.sql.functions import array, lit, col, when

df_str_01 = spark.range(6).select(
    "id",
    when(col("id") == 0, array(lit("Spark"), lit("Python"), lit("SQL")))
    .when(col("id") == 1, array(lit("Scala"), lit("R"), lit("Java")))
    .when(col("id") == 2, array(lit("Go"), lit("Rust"), lit("Julia")))
    .when(col("id") == 3, array(lit("C++"), lit("C#"), lit("F#")))
    .when(col("id") == 4, array(lit("Ruby"), lit("Perl"), lit("PHP")))
    .otherwise(array(lit("Swift"), lit("Kotlin"), lit("TypeScript")))
    .alias("skills")
)

df_str_01.printSchema()
display(df_str_01)
```

```python
# How to define schema for ArrayType
data = [
    (1, ["bajaj", "airforce", "check"]),
    (2, ["dolphin", "catia", "edge"]),
    (3, ["anand", "basketball", "orchid"]),
    (4, ["adarsh", "almond", "orphan"]),
    (5, ["superman", None]),
    (6, ["", None]),
    (7, ["diamond", "apple"]),
    (8, ["", ""])
]

schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("words", ArrayType(StringType()), True)
])

df_str_02 = spark.createDataFrame(data, schema)
display(df_str_02)
df_str_02.printSchema()
```

##### 5) Create array from existing columns

```python
from pyspark.sql.functions import array

data = [
    (1, "Java", "Python", "Spark"),
    (2, "Scala", "R", "SQL"),
    (3, "Go", "Rust", "Julia"),
    (4, "C++", "C#", "PHP"),
    (5, "Ruby", "Perl", "PHP"),
    (6, "Swift", "Kotlin", "TypeScript")
]

columns = ["id", "s1", "s2", "s3"]

df_arry = spark.createDataFrame(data, columns)
display(df_arry)
```

```python
df_arry_01 = df_arry.select(
    "*",
    array("s1", "s2", "s3").alias("skills")
)

display(df_arry_01)
df_arry_01.printSchema()
```

##### 6) Create array while creating DataFrame

```python
data = [("Naresh", ["databricks", "Azure", "SQL"], [5, 7, 8], [True, False, True], [10.5, 12.5, 15.5]),
        ("Kalmesh", ["ADF", "AWS", "Spark", "SQL"], [6, 2, 3, 12], [True, False, True, False], [10.5, 12.5, 15.5, 17.5]),
        ("Rohit", ["GCC", "PySpark", "Devops"], [3, 6, 8], [True, False, True], [10.5, 12.5, 15.5, 17.5]),
        ("Kumar", ["GitHub", "PowerBi", "Tableau", "SQL"], [2, 5, 6, 10], [True, False, True, False], [10.5, 12.5, 15.5, 17.5]),
        ("Ravi", ["Databricks", "Azure", "SQL"], [5, 7, 8], [True, False, True], [10.5, 12.5, 15.5]),
        ("Rohini", ["SQL Editor", "Python", "Oracle", "Azure", "AWS"], [4, 7, 3, 8, 9], [True, False, True, False, True], [10.5, 12.5, 15.5, 17.5, 19.5])
       ]

columns = ["Name", "Technology", "Experience", "Flags", "Prices"]

df_str_int = spark.createDataFrame(data=data, schema=columns)
display(df_str_int)
df_str_int.printSchema()
```

```python
df_schema = spark.createDataFrame(data=data, schema=schema)
display(df_schema)
df_schema.printSchema()
```

##### 7) Use STRUCT instead of ARRAY
- If you want **different data types** together, use **struct**.

```python
from pyspark.sql.functions import struct

df_struct = spark.range(6).select("id",
    struct(
        lit("Spark").alias("tool"),
        lit(10).alias("count"),
        lit(True).alias("active"),
        lit(3.14).alias("version")
    ).alias("details")
)

df_struct.printSchema()
display(df_struct)
```

##### 8) Array of STRUCTs

```python
df_arry_struct = spark.range(6).select("id",
    array(
        struct(lit("Spark").alias("name"), lit(10).alias("count")),
        struct(lit("Python").alias("name"), lit(20).alias("count"))
    ).alias("tools")
)

df_arry_struct.printSchema()
display(df_arry_struct)
```

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, ArrayType
from pyspark.sql.functions import struct

data = [
    (1, [{"name": "A", "score": 90}, {"name": "B", "score": 85}]),
    (2, [{"name": "C", "score": 70}, {"name": "D", "score": 95}]),
    (3, [{"name": "E", "score": 90}, {"name": "F", "score": 85}]),
    (4, [{"name": "G", "score": 70}, {"name": "H", "score": 95}]),
    (5, [{"name": "I", "score": 90}, {"name": "J", "score": 85}]),
    (6, [{"name": "K", "score": 70}, {"name": "L", "score": 95}]),
    (7, [{"name": "M", "score": 70}, {"name": "N", "score": 95}]),
    (8, [{"name": "O", "score": 90}, {"name": "P", "score": 85}]),
    (9, [{"name": "Q", "score": 70}, {"name": "R", "score": 95}]),
    (10, [{"name": "S", "score": 90}, {"name": "T", "score": 85}])
]

schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("students", ArrayType(StructType([
        StructField("name", StringType(), True),
        StructField("score", IntegerType(), True)
    ])), True)
])

df_arr_struct = spark.createDataFrame(data, schema)
display(df_arr_struct)
df_arr_struct.printSchema()
```

| Requirement            | Use               |
| ---------------------- | ----------------- |
| Same data type         | `array()`         |
| Different data types   | `struct()`        |
| Repeated mixed objects | `array<struct>`   |
