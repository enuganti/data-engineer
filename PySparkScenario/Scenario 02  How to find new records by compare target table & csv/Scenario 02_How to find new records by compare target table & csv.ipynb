**Scenario:** 
- I have `target table` with full load.
- Now wants to insert `only new records` from `new dataset (.csv)` by comparing with `existing target table`.

#### **Method 01**

```
%sql
-- Create a table from fullload_uat.csv
CREATE OR REPLACE TABLE full_load_table AS
SELECT
  *
FROM
  read_files(
    '/FileStore/tables/initload.csv',
    format => 'csv',
    header => true,
    mode => 'FAILFAST'
  )
```

```
%sql
SELECT * FROM full_load_table
```

```
%sql
-- Read incrementalload.csv
CREATE OR REPLACE TEMP VIEW incremental_load_view AS
SELECT
  *
FROM
  read_files(
    '/FileStore/tables/newload.csv',
    format => 'csv',
    header => true,
    mode => 'FAILFAST'
  )
```

```
%sql
SELECT * FROM incremental_load_view
```

```
%sql
-- Insert only new records into the table
INSERT INTO full_load_table
SELECT * FROM incremental_load_view ilv
WHERE NOT EXISTS (
  SELECT * FROM full_load_table flt
  WHERE flt.Cust_Id = ilv.Cust_Id
);
```

```
%sql
SELECT * FROM full_load_table
```

#### **Method 02**

```
df_trg = spark.read.csv("/FileStore/tables/initload.csv", header=True, inferSchema=True)
display(df_trg.limit(10))
print("Number of Rows:", df_trg.count())
```

```
# Assuming df_trg is your DataFrame
df_trg.createOrReplaceTempView("init_load_view")

# Now you can run SQL queries against this temporary view
result_df_trg = spark.sql("SELECT * FROM init_load_view")
display(result_df_trg)
```

```
%sql
CREATE OR REPLACE TABLE trg_load_table AS
SELECT * FROM init_load_view
```

```
df_incr = spark.read.csv("/FileStore/tables/newload.csv", header=True, inferSchema=True)
display(df_incr.limit(10))
print("Number of Rows:", df_incr.count())
```

```
# Assuming df_trg is your DataFrame
df_incr.createOrReplaceTempView("new_load_view")

# Now you can run SQL queries against this temporary view
result_df_incr = spark.sql("SELECT * FROM new_load_view")
display(result_df_incr)
```

```
%sql
CREATE OR REPLACE TABLE new_load_table AS
SELECT * FROM new_load_view
```

```
%sql

INSERT INTO trg_load_table
SELECT
  Company_Name,
  Cust_Id,
  Cust_Name,
  Category,
  Start_Date,
  Start_Cust_Date,
  End_Date,
  Updated_Date,
  Cust_Value,
  Cust_Type,
  Exchange,
  Location,
  Last_Date_UTC,
  Cust_Category,
  Index
FROM
  new_load_table
WHERE
  NOT EXISTS (
    SELECT
      *
    FROM
      trg_load_table t
    WHERE
      new_load_table.Cust_Id = t.Cust_Id
  );
```

#### **Method 03**

```
%sql
-- Create a table from initload_uat.csv
CREATE OR REPLACE TABLE targetnew_load_table AS
SELECT
  *
FROM
  read_files(
    '/FileStore/tables/initload.csv',
    format => 'csv',
    header => true,
    mode => 'FAILFAST'
  )
```

```
%sql
SELECT * FROM targetnew_load_table
```

```
%sql
INSERT INTO
  targetnew_load_table (
    Company_Name,
    Cust_Id,
    Cust_Name,
    Category,
    Start_Date,
    Start_Cust_Date,
    End_Date,
    Updated_Date,
    Cust_Value,
    Cust_Type,
    Exchange,
    Location,
    Last_Date_UTC,
    Cust_Category,
    Index
    ) WITH new_load AS (
    SELECT
      *
    FROM
      read_files(
        '/FileStore/tables/newload.csv',
        format => 'csv',
        header => true,
        mode => 'FAILFAST'
      )
  )
SELECT
  Company_Name,
  Cust_Id,
  Cust_Name,
  Category,
  Start_Date,
  Start_Cust_Date,
  End_Date,
  Updated_Date,
  Cust_Value,
  Cust_Type,
  Exchange,
  Location,
  Last_Date_UTC,
  Cust_Category,
  Index
FROM
  new_load
WHERE
  NOT EXISTS (
    SELECT
      *
    FROM
      targetnew_load_table t
    WHERE
      new_load.Cust_Id = t.Cust_Id
  )
```

```
%sql
SELECT * FROM targetnew_load_table
```
