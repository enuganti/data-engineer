**How to access Nested dictionary?**
- If a `value itself` is `another dictionary`, you can `chain the keys`.

**Different Methods to Access Nested Dictionary Items**

**a) Direct Key Access**
- Use **square brackets** repeatedly:

```python
student = {
    "name": "Ravi",
    "marks": {
        "math": 90,
        "science": 85,
        "languages": {
            "english": 88,
            "hindi": 92
        }
    }
}
```

```python
score = student["marks"]["science"]
print("Student Marks in Science: ", score)
```

```python
# Accessing directly
print("Student Marks in Math: ", student["marks"]["math"])
print("Student Marks in Hindi: ", student["marks"]["languages"]["hindi"])
```

```python
data = {
    "student1": {"name": "Rahul", "age": 20},
    "student2": {"name": "Priya", "age": 22}
}

print("Name of Student1: ", data["student1"]["name"])
print("Name of Student2: ", data["student2"]["name"])
print("Age of Student1: ", data["student1"]["age"])
print("Age of Student2: ", data["student2"]["age"])
```

**b) Using .get()**

```python
student = {
    "name": "Ravi",
    "marks": {
        "math": 90,
        "science": 85,
        "languages": {
            "english": 88,
            "hindi": 92
        }
    }
}

print("Name of Student: ", student.get("name"))
print("Marks of Student: ", student.get("marks"))
print("Languages of Student: ", student.get("languages"))
print("Languages of Student: ", student.get("languages"), {})
print("Languages of Student: ", student.get("marks", {}).get("languages"))
print("Languages of Student: ", student.get("marks", {}).get("score", {})) # dict.get("key", default_value)
```

```python
student.get("marks", {})
```

- Looks for key `"marks"`.
  - `Found → returns`
  - `If not found` → returns `empty {}` (safe fallback)

```python
L1 = student.get("marks", {})
L1.get("languages", {})
```

- Looks for `language inside marks`.
  - `Found → returns`
  - `If missing → returns {}`

```python
L2 = L1.get("languages", {})
L2.get("english")
```

- Looks `inside languages`.
  - `Found → returns`
  - `If missing` → returns `None`

```python
score = student.get("marks", {}).get("languages", {}).get("english")
print("Student Marks in English: ", score)
```

**c) Looping Through Nested Dictionaries**

```python
for subject, mark in student["marks"].items():
    print(subject, ":", mark)
```

**If the value is another dictionary**

```python
for lang, mark in student["marks"]["languages"].items():
    print(lang, ":", mark)
```

**d) Using Recursion (for unknown depth)**

```python
student = {
    "name": "Ravi",
    "age": 25,
    "gender": "male",
    "marks": {
        "math": 90,
        "science": 85,
        "computer": 78,
        "languages": {
            "english": 88,
            "hindi": 92,
            "telugu": 95,
            "tamil": 85,
            "kannada": 80
        }
    }
}
```

```python
def print_dict(d):
    for key, value in d.items():
        if isinstance(value, dict):  # Checking if the value is another dictionary; If YES → go inside it; If NO → print it
            print_dict(value)
        else:
            print(key, ":", value)

print_dict(student)
```
