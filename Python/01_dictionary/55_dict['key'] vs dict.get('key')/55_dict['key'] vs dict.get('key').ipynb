##### dict['key'] vs dict.get('key') for Accessing values from Dictionary

**1) Direct Key Access**
- Use **square brackets** repeatedly:
- `❌ Normal access (can break)`
- `❌ Error`
  - `KeyError: 'price'`

```python
sales = {
    "name": "Karishma",
    "gender": "female",
    "age": 25,
    "products": {
        "samsung": 100,
        "apple": 200,
        "nokia": 90,
        "motorola": 150,
        "purchase": {
            "sold_date": "2022-01-01",
            "price": 92,
            "discount": 10
        }
    }
}
```

```python
# Accessing directly
print("Name of the Sales Person: ", sales["name"])    # Dict["key"]
print("Number of samsung products: ", sales["products"]["samsung"])
print("Number of nokia products: ", sales["products"]["nokia"])
print("Product purchase date: ", sales["products"]["purchase"]["sold_date"])
print("Product discount: ", sales["products"]["purchase"]["discount"])
```

```python
# Using [] results in KeyError
print(sales['Marketing'])
```

**2) Safe access using get()**

```python
print(sales.get("name"))   # Dict.get("key")
```

```python
print(sales.get("Marketing"))
```

```python
print(sales.get("Marketing", "Marketing not found"))
```

- **Safer than []**, since `get` method won’t **throw an error** if the **key is missing**.
