Python join() is an inbuilt string function used to **join elements** of a sequence **separated by a string separator**. This function joins elements of a sequence and makes it a string.

**Syntax:**

     separator_string.join(iterable)

**Parameters:**

- **Iterable:** objects capable of returning their members one at a time. Some examples are **List, Tuple, String, Dictionary, and Set**.

- **Return Value:** The join() method returns a string concatenated with the elements of iterable.

- **Type Error:** If the iterable contains any **non-string** values, it raises a **TypeError exception**.

**1) Joining substrings in a string**

```python
my_string = "python"
separator = ' '
result = separator.join(my_string)
print(result)
```

```python
# This will join the characters of the string 'hello' with '-'
str = '-'.join('hello')
print(str)
```

**2) How to Join a List into a String?**

```python
list_abc = ['aaa', 'bbb', 'ccc']
string = '-'.join(list_abc)
print(string)
```

```python
# join all the elements in the list using an empty string as a separator
list1 = ['g', 'e', 'e', 'k', 's']
print("".join(list1))
```

```python
# join the elements of the list using “$” as a separator
list1 = " geeks "
print("$".join(list1))
```

**3) How to Joining a list of Strings with a Custom Separator?**

```python
words = ["apple", "", "banana", "cherry", ""]
separator = "@ "
result = separator.join(word for word in words if word)
print(result)
```

```python
# Joining elements with a custom delimiter
my_list = ['apple', 'banana', 'orange']
delimiter = ' -> '
result = delimiter.join(my_list)
print(result)
```

```python
# How to join elements of a list with a space as a separator?
list_of_strings = ['Hello', 'World']
result = ' '.join(list_of_strings)
print(result)
```

```python
# How to join elements of a list with a comma as a separator?
list_of_strings = ['apple', 'banana', 'cherry']
result = ','.join(list_of_strings)
print(result)
```

```python
list_abc = ['aaa', 'bbb', 'ccc']
string = '\n'.join(list_abc)
print(string)
```
